```python
####  I PREPROCESAMIENTO PROG DETALLADA ######

## 1.-
### REDUCCIÓN DE COLUMNAS REQUERIDAS###########################
columnas_requeridas = [
    'PERIODO', 'NRC', 'MATERIA', 'N_CURSO', 'SECCION',
    'COMPONENTE', 'CAMPUS', 'NOMBRE', 'VACANTES', 'INSCRITOS', 'EDIFICIO',
    'SALA', 'CAPACIDAD', 'FECHA_INI2', 'FECHA_TERM2', 'HR_INICIO', 'HR_FIN',
    'LUNES', 'MARTES', 'MIERCOLES', 'JUEVES', 'VIERNES', 'SABADO', 'DOMINGO'
]

# Filtrar el DataFrame manteniendo solo las columnas existentes
df = df[[col for col in columnas_requeridas if col in df.columns]]

## 2.-
## FUERA DUPLICADOS!!!!####
df.drop_duplicates(inplace=True)

##3.-
##########ESTANDARIZACIÓN DE LOS BLOQUES########################

PROG_DETALLADA = 'C:/Users/diego.perezp/Downloads/MAPA_CALOR_09122025/PROG_DETALLADA_01122025_COMPLETA_2025.csv'

# Paso 1: Convertir las horas de df2 a formato numérico HHMM para facilitar la comparación
def hora_str_a_numero(hora_str):
    """Convierte 'HH:MM:SS' a HHMM como entero"""
    try:
        h, m, s = hora_str.split(':')
        return int(h) * 100 + int(m)
    except:
        return None

# Aplicar conversión a df2
df2['HORA_INICIO_NUM'] = df2['HORA INICIO'].apply(hora_str_a_numero)
df2['HORA_FIN_NUM'] = df2['HORA FIN'].apply(hora_str_a_numero)

# Crear listas ordenadas de los tiempos de bloque
bloques_inicio = sorted(df2['HORA_INICIO_NUM'].dropna().unique())
bloques_fin = sorted(df2['HORA_FIN_NUM'].dropna().unique())

# Paso 2: Función para encontrar el tiempo más cercano en los bloques
def encontrar_mas_cercano(hora_actual, bloques):
    """Encuentra el bloque más cercano a la hora actual"""
    if pd.isna(hora_actual):
        return None
    
    # Convertir a entero si es string
    if isinstance(hora_actual, str):
        try:
            # Eliminar : si existe
            hora_actual = hora_actual.replace(':', '')
            hora_num = int(hora_actual[:4])  # Tomar solo HHMM
        except:
            return None
    else:
        hora_num = int(hora_actual)
    
    # Buscar el bloque más cercano
    dif_min = float('inf')
    bloque_cercano = None
    
    for bloque in bloques:
        dif = abs(hora_num - bloque)
        if dif < dif_min:
            dif_min = dif
            bloque_cercano = bloque
    
    return bloque_cercano

# Paso 3: Aplicar las funciones a df
df['HR_INICIO_AUX'] = df['HR_INICIO'].apply(
    lambda x: encontrar_mas_cercano(x, bloques_inicio)
)
df['HR_FIN_AUX'] = df['HR_FIN'].apply(
    lambda x: encontrar_mas_cercano(x, bloques_fin)
)

# Paso 4: Convertir de vuelta a formato string HH:MM para consistencia
def numero_a_hora_str(numero):
    """Convierte HHMM entero a 'HH:MM'"""
    if pd.isna(numero):
        return None
    numero = int(numero)
    horas = numero // 100
    minutos = numero % 100
    return f"{horas:02d}:{minutos:02d}"

df['HR_INICIO_AUX_STR'] = df['HR_INICIO_AUX'].apply(numero_a_hora_str)
df['HR_FIN_AUX_STR'] = df['HR_FIN_AUX'].apply(numero_a_hora_str)
##############################################################################

## 4.-
### VARIABLE AUX AÑO #########################################################
df['AÑO'] = df['PERIODO'].astype(str).str[:4].astype(int)

## 5.-
####### CAMBIO DE NAN A VACIO ################################################
df = df.fillna('')

## 6.-
## ELIMINAR REGISTROS DE PROGRAMACIÓN SIN HORA DE INICIO Y/O FIN #############
df = df[~df['HR_INICIO'].isna() & (df['HR_INICIO'] != '') & (df['HR_FIN'] != '') & (df['HR_FIN'] != '')]
```

```python
####  II PREPROCESAMIENTO PROG DETALLADA ######

## 1.- AUX DF ###############################################################
df_separado = df

## 2.- 
###SEPARACIÓN DE DÍAS 1 X LINEA DE REGISTRO #################################

# Primero, verificar si ya tenemos la columna DIA_SESION
if 'DIA_SESION' in df_separado.columns:
    print("-----------------------")
    
    # Verificar si hay filas con múltiples días
    filas_con_multiples_dias = 0
    for _, fila in df_separado.iterrows():
        dias_marcados = [dia for dia in columnas_dias if pd.notna(fila[dia]) and str(fila[dia]).strip() != '']
        if len(dias_marcados) > 1:
            filas_con_multiples_dias += 1
    
    print(f"Filas con múltiples días: {filas_con_multiples_dias}")
    
    if filas_con_multiples_dias == 0:
        print("Todos los registros ya están separados por día. No es necesario reprocesar.")
    else:
        print("Reprocesando para separar días...")
else:
    print("Creando separación por días...")

# Lista de columnas de días de la semana
columnas_dias = ['LUNES', 'MARTES', 'MIERCOLES', 'JUEVES', 'VIERNES', 'SABADO', 'DOMINGO']

# Crear una lista para almacenar las nuevas filas
filas_separadas = []

# Recorrer cada fila del DataFrame ORIGINAL (df, no df_separado)
for _, fila in df.iterrows():  # ¡CORRECCIÓN: usar df, no df_separado!
    # Verificar cuáles días están marcados (no vacíos)
    dias_marcados = [dia for dia in columnas_dias if pd.notna(fila[dia]) and str(fila[dia]).strip() != '']
    
    # Si no hay días marcados, mantener la fila como está
    if not dias_marcados:
        fila_copia = fila.copy()
        fila_copia['DIA_SESION'] = ''
        filas_separadas.append(fila_copia)
    else:
        # Crear una fila por cada día marcado
        for dia in dias_marcados:
            nueva_fila = fila.copy()
            
            # Mantener solo el día actual en las columnas de días
            for col_dia in columnas_dias:
                if col_dia != dia:
                    # Vaciar los demás días
                    nueva_fila[col_dia] = ''
            
            # Agregar columna para identificar el día de la sesión
            nueva_fila['DIA_SESION'] = dia
            
            filas_separadas.append(nueva_fila)

# Crear un nuevo DataFrame con las filas separadas
df_separado = pd.DataFrame(filas_separadas)

# Opcional: reordenar las columnas para que DIA_SESION esté cerca de las columnas de días
columnas_ordenadas = list(df.columns)  # Usar columnas del df original
# Insertar DIA_SESION después de la última columna de día
if 'DOMINGO' in columnas_ordenadas:
    indice_insertar = columnas_ordenadas.index('DOMINGO') + 1
else:
    # Si no existe DOMINGO, buscar la última columna de día que exista
    for dia in reversed(columnas_dias):
        if dia in columnas_ordenadas:
            indice_insertar = columnas_ordenadas.index(dia) + 1
            break
    else:
        indice_insertar = len(columnas_ordenadas)

# Agregar DIA_SESION a la lista de columnas ordenadas
if 'DIA_SESION' not in columnas_ordenadas:
    columnas_ordenadas.insert(indice_insertar, 'DIA_SESION')

# Asegurar que todas las columnas existan en df_separado
columnas_finales = [col for col in columnas_ordenadas if col in df_separado.columns]
df_separado = df_separado[columnas_finales]

print(f"=== RESULTADO ===")
print(f"Filas originales en df: {len(df)}")
print(f"Filas después de separar en df_separado: {len(df_separado)}")


#############################################################################

## 3.-
## Eliminar filas donde DIA_SESION sea NaN o esté vacío (string vacío) #########################
df_separado = df_separado[~df_separado['DIA_SESION'].isna() & (df_separado['DIA_SESION'] != '')]
```

```python
####  III PREPROCESAMIENTO PROG DETALLADA ######

## 1.-
####### APERTURA POR BOLOQUE HORARIO #####################################

# Mostrar cantidad de registros originales
print(f"Registros originales en df_separado: {len(df_separado)}")

# Convertir las columnas de tiempo a datetime para comparación
df2['HORA INICIO'] = pd.to_datetime(df2['HORA INICIO']).dt.time
df2['HORA FIN'] = pd.to_datetime(df2['HORA FIN']).dt.time

# Convertir las columnas de tiempo en df_separado a datetime
df_separado['HR_INICIO_AUX_STR'] = pd.to_datetime(df_separado['HR_INICIO_AUX_STR']).dt.time
df_separado['HR_FIN_AUX_STR'] = pd.to_datetime(df_separado['HR_FIN_AUX_STR']).dt.time

# Lista para almacenar los registros aperturados
registros_aperturados = []

# Iterar sobre cada fila del df_separado
for _, row in df_separado.iterrows():
    # Obtener el tiempo de inicio y fin de la sesión
    inicio_sesion = row['HR_INICIO_AUX_STR']
    fin_sesion = row['HR_FIN_AUX_STR']
    
    # Filtrar bloques que estén dentro del rango de la sesión
    bloques_sesion = df2[
        (df2['HORA INICIO'] >= inicio_sesion) & 
        (df2['HORA FIN'] <= fin_sesion)
    ]
    
    # Si no hay bloques exactos, buscar bloques que se solapen
    if len(bloques_sesion) == 0:
        # Buscar bloques que empiecen después o igual al inicio y terminen antes o igual al fin
        bloques_sesion = df2[
            (df2['HORA INICIO'] >= inicio_sesion) & 
            (df2['HORA FIN'] <= fin_sesion)
        ]
    
    # Para cada bloque en la sesión, crear un nuevo registro
    for _, bloque in bloques_sesion.iterrows():
        nuevo_registro = row.copy()
        
        # Actualizar las horas con las del bloque estándar
        nuevo_registro['HR_INICIO_AUX'] = int(bloque['HORA INICIO'].strftime('%H%M'))
        nuevo_registro['HR_FIN_AUX'] = int(bloque['HORA FIN'].strftime('%H%M'))
        nuevo_registro['HR_INICIO_AUX_STR'] = bloque['HORA INICIO'].strftime('%H:%M:%S')
        nuevo_registro['HR_FIN_AUX_STR'] = bloque['HORA FIN'].strftime('%H:%M:%S')
        nuevo_registro['BLOQUE'] = bloque['BLOQUE']
        
        # Agregar el nuevo registro a la lista
        registros_aperturados.append(nuevo_registro)
        
        # Si llegamos al bloque que coincide con el fin de la sesión, detenemos la iteración
        if bloque['HORA FIN'] == fin_sesion:
            break

# Crear el DataFrame aperturado
df_aperturado = pd.DataFrame(registros_aperturados)

# Reordenar columnas para que BLOQUE esté en la posición correcta
columnas = list(df_aperturado.columns)
if 'BLOQUE' in columnas:
    columnas.remove('BLOQUE')
    # Insertar BLOQUE después de DIA_SESION
    if 'DIA_SESION' in columnas:
        idx = columnas.index('DIA_SESION') + 1
        columnas.insert(idx, 'BLOQUE')
    else:
        columnas.insert(0, 'BLOQUE')

df_aperturado = df_aperturado[columnas]

# Mostrar cantidad de registros después de aperturar
print(f"Registros después de aperturar: {len(df_aperturado)}")
print(f"Incremento: {len(df_aperturado) - len(df_separado)} registros ({((len(df_aperturado) / len(df_separado)) - 1) * 100:.1f}% más)")
########################################################################################################

## 2.-
### VARIABLE LLAV2################################################### 
campos_para_llav2 = ['PERIODO', 'NRC', 'DIA_SESION', 'HR_INICIO_AUX']

# Concatenación  (todos como strings)
df_separado['LLAV2'] = df_separado[campos_para_llav2].astype(str).agg(''.join, axis=1)

## 3.-
### VARIABLE LLAV2###### ############################################
campos_para_llav2 = ['PERIODO', 'NRC', 'DIA_SESION', 'HR_INICIO_AUX']

# Opción 1: Concatenación simple (todos como strings)
df_aperturado['LLAV2'] = df_aperturado[campos_para_llav2].astype(str).agg(''.join, axis=1)

## 4.-
## Marcar con 1 el primer registro único de cada LLAV2, 0 para los duplicados ###########################
df_aperturado['REG_UNICO_LLAV2'] = (~df_aperturado.duplicated(subset=['LLAV2'])).astype(int)
```

```python
 #### PREPROCESAMIENTO ESPACIO FISICO ###############################

## 1.-
### CAMBIO DE NAN A VACIO ##########################
df3 = df3.fillna(0)

## 2 .-
####### VARIABLE LLAV3###### para sacar el registro más actualizado ###
campos_para_llav3 = ['Campus Edificio', 'Código Edificio', 'Número Sala']

#######  Concatenación simple (todos como strings) ###################
df3['LLAV3'] = df3[campos_para_llav3].astype(str).agg(''.join, axis=1)

## 3.-
##### CONVERSIÓN DE PERIODO A DATO NÚMERICO + DETECCIÓN DE REGISTRO MÁS ACTUAL ########
df3['Periodo'] = pd.to_numeric(df3['Periodo'], errors='coerce')

# Encontrar el índice del periodo máximo para cada grupo LLAV3
idx_max = df3.groupby('LLAV3')['Periodo'].idxmax()

# Crear nueva columna 'Es_Mas_Actualizado' con 1 para los registros más actualizados
df3['Es_Mas_Actualizado'] = 0
df3.loc[idx_max, 'Es_Mas_Actualizado'] = 1

## 4.-
### Mapeo de salas ##########################################################################
# Primero, definimos el diccionario de mapeo para las reglas
reglas = {
    'Auditorio': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'ESPACIO USS'},
    'CAMPO CLINICO': {'EXTERNO': 'SI', 'CLASIFICACIÓN': 'EXTERNO'},
    'Campo Clínico': {'EXTERNO': 'SI', 'CLASIFICACIÓN': 'EXTERNO'},
    'Centro de Práctica': {'EXTERNO': 'SI', 'CLASIFICACIÓN': 'EXTERNO'},
    'Centro de Salud': {'EXTERNO': 'SI', 'CLASIFICACIÓN': 'EXTERNO'},
    'Espacio Educativo USS': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'ESPACIO USS'},
    'Gimnasio': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'ESPACIO USS'},
    'Hospital de Simulación': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'LABORATORIO'},
    'Laboratorio': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'LABORATORIO'},
    'Laboratorio de Computación': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'LABORATORIO'},
    'Otros': {'EXTERNO': 'SI', 'CLASIFICACIÓN': 'EXTERNO'},
    'Sala de Clases': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'SALA'},
    'Sala de Clases Híbrida': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'SALA'},
    'Sala de Clases Virtual': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'SALA VIRTUAL'},
    'Sala de estudio': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'ESPACIO USS'},
    'Sala de reuniones': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'ESPACIO USS'},
    'Sala Espejo': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'LABORATORIO'},
    'Taller': {'EXTERNO': 'NO', 'CLASIFICACIÓN': 'ESPACIO USS'}
}

# Función para aplicar las reglas
def aplicar_reglas(categoria_sala):
    if pd.isna(categoria_sala) or str(categoria_sala).strip() == '':
        return 'SI', 'EXTERNO'
    
    categoria_str = str(categoria_sala).strip()
    
    if categoria_str in reglas:
        return reglas[categoria_str]['EXTERNO'], reglas[categoria_str]['CLASIFICACIÓN']
    else:
        # Para categorías no definidas en las reglas
        return 'SI', 'EXTERNO'

# Aplicar la función a la columna 'Categoría Sala'
df3[['EXTERNO', 'CLASIFICACIÓN']] = df3['Categoría Sala'].apply(
    lambda x: pd.Series(aplicar_reglas(x))
)
## 5.-
### Clasificación de Rango_Capacidad ##########
df3['Rango_Capacidad'] = pd.cut(
    df3['Capacidad Máxima Sala'],
    bins=[0, 20, 40, 60, float('inf')],
    labels=['0-20', '21-40', '41-60', '60 o más'],
    right=True
)
```

```python
####  IV PREPROCESAMIENTO PROG DETALLADA ######
## 1.-
####### CARGA DE CLASIFICACION DE ESPACIO FISICO EN PROG DETALLADA ###############

# Crear una clave de cruce en df3 (Código Edificio + Número Sala)
df3['clave_cruce'] = df3['Código Edificio'].astype(str) + df3['Número Sala'].astype(str)

# Crear una clave de cruce en df_aperturado (EDIFICIO + SALA)
df_aperturado['clave_cruce'] = df_aperturado['EDIFICIO'].astype(str) + df_aperturado['SALA'].astype(str)

# Crear un diccionario de mapeo desde df3 (usando el registro más actualizado si hay duplicados)
# Primero filtramos por Es_Mas_Actualizado = 1 si existen múltiples registros
if 'Es_Mas_Actualizado' in df3.columns:
    df3_actualizado = df3[df3['Es_Mas_Actualizado'] == 1]
else:
    df3_actualizado = df3
    
# Eliminar duplicados manteniendo el primero (o el más reciente si ya está ordenado)
mapeo_clasificacion = df3_actualizado.drop_duplicates(subset='clave_cruce', keep='first')
mapeo_dict = mapeo_clasificacion.set_index('clave_cruce')['CLASIFICACIÓN'].to_dict()

# Mapear la clasificación a df_aperturado usando la clave de cruce
df_aperturado['CLASIFICACIÓN'] = df_aperturado['clave_cruce'].map(mapeo_dict)

# Rellenar valores no encontrados con "Sin Clasificación"
df_aperturado['CLASIFICACIÓN'] = df_aperturado['CLASIFICACIÓN'].fillna('Sin Clasificación')

# **NUEVO: Mapear el campo Campus Edificio como CAMPUS_AUX**
mapeo_campus = mapeo_clasificacion.set_index('clave_cruce')['Campus Edificio'].to_dict()
df_aperturado['CAMPUS_AUX'] = df_aperturado['clave_cruce'].map(mapeo_campus)

# Si no se encuentra el cruce, usar el valor de CAMPUS
df_aperturado['CAMPUS_AUX'] = df_aperturado['CAMPUS_AUX'].fillna(df_aperturado['CAMPUS'])

# **NUEVO: Mapear el campo Capacidad Máxima Sala como Capacidad Máxima Sala_pd1**
mapeo_capacidad = mapeo_clasificacion.set_index('clave_cruce')['Capacidad Máxima Sala'].to_dict()
df_aperturado['Capacidad Máxima Sala_pd1'] = df_aperturado['clave_cruce'].map(mapeo_capacidad)

# Si no se encuentra el cruce, usar el valor de CAPACIDAD
df_aperturado['Capacidad Máxima Sala_pd1'] = df_aperturado['Capacidad Máxima Sala_pd1'].fillna(df_aperturado['CAPACIDAD'])

# **NUEVO: Mapear el campo Rango_Capacidad**
mapeo_rango = mapeo_clasificacion.set_index('clave_cruce')['Rango_Capacidad'].to_dict()
df_aperturado['Rango_Capacidad'] = df_aperturado['clave_cruce'].map(mapeo_rango)
# Si no se encuentra el cruce, completar con "Sin Capacidad"
df_aperturado['Rango_Capacidad'] = df_aperturado['Rango_Capacidad'].fillna('Sin Capacidad')
# Opcional: eliminar la columna temporal de clave de cruce si no la necesitas
df_aperturado = df_aperturado.drop(columns=['clave_cruce'])
df3 = df3.drop(columns=['clave_cruce'])

## 2.- 
##### CARGA DE DESCRIPTORES EN PROG DETALLADA  ################
# Hacer merge entre df_aperturado y df4
df_aperturado = pd.merge(
    df_aperturado,
    df4[['stvsubj_code', 'stvsubj_desc']],
    left_on='MATERIA',
    right_on='stvsubj_code',
    how='left'
)
# Eliminar la columna si ya existe antes de renombrar
if 'DESCRIPCIÓN CARRERA_MATERIA' in df_aperturado.columns:
    df_aperturado = df_aperturado.drop(columns=['DESCRIPCIÓN CARRERA_MATERIA'])
    
# Renombrar la columna resultante
df_aperturado = df_aperturado.rename(columns={'stvsubj_desc': 'DESCRIPCIÓN CARRERA_MATERIA'})
# Opcional: eliminar la columna stvsubj_code si no la necesitas
df_aperturado = df_aperturado.drop(columns=['stvsubj_code'])

## 3.-
##### CALCULO DE OCUPACIÓN_SALA ################
# Convertir Capacidad Máxima Sala_pd1 a numérico (por si viene como string)
df_aperturado['Capacidad Máxima Sala_pd1'] = pd.to_numeric(df_aperturado['Capacidad Máxima Sala_pd1'], errors='coerce')
# Calcular OCUPACIÓN_SALA = INSCRITOS / Capacidad Máxima Sala_pd1
# Si no hay capacidad (0 o NaN), dejar en 0
# Si el resultado es > 1, truncar en 1
df_aperturado['OCUPACIÓN_SALA'] = df_aperturado.apply(
    lambda row: 0 if pd.isna(row['Capacidad Máxima Sala_pd1']) or row['Capacidad Máxima Sala_pd1'] == 0 
    else min(row['INSCRITOS'] / row['Capacidad Máxima Sala_pd1'], 1),
    axis=1
)
```
