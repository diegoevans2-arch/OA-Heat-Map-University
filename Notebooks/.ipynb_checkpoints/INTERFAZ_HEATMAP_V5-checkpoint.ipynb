```python

########## FUNCIÓN PARA GENERAR PDF DE RESUMEN EJECUTIVO V2 ##########

def generar_resumen_ejecutivo_pdf(df_filtrado, df3_ref, df2_ref, filtros_dict, tipo_grafico="Heatmap"):
    """
    Genera un PDF con formato de Resumen Ejecutivo de ocupación.
    
    Parámetros:
    - df_filtrado: DataFrame con los datos filtrados de ocupación
    - df3_ref: DataFrame de capacidad de salas (df3)
    - df2_ref: DataFrame de bloques estándar (df2)
    - filtros_dict: dict con los filtros aplicados
    - tipo_grafico: str - "Heatmap" o "Sunburst"
    """
    
    if df_filtrado is None or df_filtrado.empty:
        print("⚠️ No hay datos filtrados. Primero genera el gráfico.")
        return
    
    timestamp = pd.Timestamp.now().strftime('%Y%m%d_%H%M%S')
    filename = f'Resumen_Ejecutivo_{tipo_grafico}_{timestamp}.pdf'
    
    # =============================================
    # PREPARACIÓN DE DATOS
    # =============================================
    df_work = df_filtrado.copy()
    for col in ['CAMPUS_AUX', 'EDIFICIO', 'SALA', 'CLASIFICACIÓN', 'CAMPUS']:
        if col in df_work.columns:
            df_work[col] = df_work[col].astype(str)
    
    # Filtrar solo registros con SALA válida (no vacía, no null, no 'nan')
    df_work = df_work[
        (df_work['SALA'].notna()) & 
        (df_work['SALA'] != '') & 
        (df_work['SALA'] != 'nan') &
        (df_work['SALA'].str.strip() != '')
    ].copy()
    
    if df_work.empty:
        print("⚠️ No hay registros con sala válida.")
        return
    
    col_ocup = 'OCUPACIÓN_SALA' if 'OCUPACIÓN_SALA' in df_work.columns else None
    if not col_ocup:
        print("⚠️ No se encontró columna OCUPACIÓN_SALA.")
        return
    
    # --- Calcular total_capacidad del campus (desde df3 con mismos filtros) ---
    campus_edificio_list = filtros_dict.get('campus_edificio_list', [])
    excl_cap = filtros_dict.get('clasificacion_excluir_capacidad', [])
    
    filtro_df3 = df3_ref['Campus Edificio'].isin(campus_edificio_list)
    if excl_cap:
        filtro_df3 = filtro_df3 & (~df3_ref['CLASIFICACIÓN'].isin(excl_cap))
    total_capacidad = df3_ref.loc[filtro_df3, 'Capacidad Máxima Sala'].sum()
    if total_capacidad == 0:
        total_capacidad = 1  # evitar division por cero
    
    # --- Tabla de % ocupación por sala (campo OCUPACIÓN_SALA promedio) ---
    group_cols = ['CAMPUS_AUX', 'EDIFICIO', 'CLASIFICACIÓN', 'SALA']
    if 'Rango_Capacidad' in df_work.columns:
        group_cols = ['CAMPUS_AUX', 'EDIFICIO', 'CLASIFICACIÓN', 'Rango_Capacidad', 'SALA']
    
    tabla_sala = df_work.groupby(group_cols).agg(
        Ocupacion_Prom=(col_ocup, 'mean'),
        N_Registros=(col_ocup, 'count')
    ).reset_index()
    tabla_sala['Ocupacion_Pct'] = (tabla_sala['Ocupacion_Prom'] * 100).round(2)
    
    # Promedio por campus = promedio simple de OCUPACIÓN_SALA de todos los registros con sala válida
    promedio_campus = df_work.groupby('CAMPUS_AUX')[col_ocup].mean().to_dict()
    promedio_campus = {k: round(v * 100, 2) for k, v in promedio_campus.items()}
    
    # Ordenar
    tabla_sala = tabla_sala.sort_values('Ocupacion_Pct', ascending=False).reset_index(drop=True)
    
    # =============================================
    # ESTILOS PDF
    # =============================================
    doc = SimpleDocTemplate(filename, pagesize=letter,
                            topMargin=1.5*cm, bottomMargin=1.5*cm,
                            leftMargin=1.5*cm, rightMargin=1.5*cm)
    
    styles = getSampleStyleSheet()
    
    style_titulo = ParagraphStyle('TituloEjecutivo', parent=styles['Title'],
                                   fontSize=18, textColor=colors.HexColor('#1a237e'),
                                   spaceAfter=6, alignment=TA_CENTER)
    
    style_subtitulo = ParagraphStyle('Subtitulo', parent=styles['Heading2'],
                                      fontSize=13, textColor=colors.HexColor('#283593'),
                                      spaceBefore=14, spaceAfter=6)
    
    style_seccion = ParagraphStyle('Seccion', parent=styles['Heading3'],
                                    fontSize=11, textColor=colors.HexColor('#1565c0'),
                                    spaceBefore=10, spaceAfter=4)
    
    style_normal = ParagraphStyle('NormalCustom', parent=styles['Normal'],
                                   fontSize=9, leading=12, alignment=TA_JUSTIFY)
    
    style_small = ParagraphStyle('Small', parent=styles['Normal'],
                                  fontSize=8, leading=10, textColor=colors.HexColor('#555555'))
    
    style_alerta = ParagraphStyle('Alerta', parent=styles['Normal'],
                                   fontSize=9, leading=12, textColor=colors.HexColor('#b71c1c'),
                                   fontName='Helvetica-Bold')
    
    style_parrafo_resumen = ParagraphStyle('ParrafoResumen', parent=styles['Normal'],
                                            fontSize=9, leading=13, alignment=TA_JUSTIFY,
                                            spaceBefore=4, spaceAfter=4)
    
    story = []
    
    # =============================================
    # ENCABEZADO
    # =============================================
    story.append(Paragraph("RESUMEN EJECUTIVO", style_titulo))
    story.append(Paragraph("Análisis de Ocupación de Espacios Académicos",
                           ParagraphStyle('Sub', parent=styles['Normal'],
                                          fontSize=12, alignment=TA_CENTER,
                                          textColor=colors.HexColor('#455a64'), spaceAfter=4)))
    
    fecha_gen = pd.Timestamp.now().strftime('%d/%m/%Y %H:%M')
    story.append(Paragraph(f"Generado: {fecha_gen} | Fuente: {tipo_grafico}",
                           ParagraphStyle('Fecha', parent=styles['Normal'],
                                          fontSize=8, alignment=TA_CENTER,
                                          textColor=colors.HexColor('#888888'), spaceAfter=8)))
    
    story.append(HRFlowable(width="100%", thickness=2, color=colors.HexColor('#1a237e')))
    story.append(Spacer(1, 10))
    
    # =============================================
    # 1. RESUMEN EJECUTIVO EN PÁRRAFO
    # =============================================
    story.append(Paragraph("1. Resumen Ejecutivo", style_subtitulo))
    
    # Construir párrafo dinámico - Solo registros donde CAMPUS == CAMPUS_AUX (campus de análisis)
    campus_list_filtro = filtros_dict.get('campus_list', [])
    anio = filtros_dict.get('anio', '-')
    
    # Para el resumen, filtrar solo registros del campus de análisis
    df_resumen = df_work[df_work['CAMPUS'] == df_work['CAMPUS_AUX']].copy() if 'CAMPUS' in df_work.columns else df_work.copy()
    
    # Datos para el resumen (solo campus de análisis: CAMPUS == CAMPUS_AUX)
    tabla_sala_resumen = tabla_sala.copy()
    if 'Rango_Capacidad' in tabla_sala_resumen.columns:
        tabla_sala_resumen = tabla_sala_resumen[tabla_sala_resumen['Rango_Capacidad'].astype(str) != 'Sin Capacidad']
    n_salas_total = tabla_sala_resumen['SALA'].nunique()
    n_criticas = len(tabla_sala_resumen[tabla_sala_resumen['Ocupacion_Pct'] <= 20])
    
    # Bloques con menor ocupación (excluyendo sab/dom, solo campus de análisis)
    df_bloques_resumen = df_resumen[~df_resumen['DIA_SESION'].isin(['SABADO', 'DOMINGO'])].copy() if 'DIA_SESION' in df_resumen.columns else df_resumen.copy()
    
    bloques_bajos_resumen = []
    if 'HR_INICIO_AUX_STR' in df_bloques_resumen.columns:
        bloq_r = df_bloques_resumen.groupby('HR_INICIO_AUX_STR')['INSCRITOS'].sum().reset_index()
        # contar días L-V por bloque
        dias_por_bloque_r = df_bloques_resumen.groupby('HR_INICIO_AUX_STR')['DIA_SESION'].nunique().reset_index()
        dias_por_bloque_r.columns = ['HR_INICIO_AUX_STR', 'N_DIAS']
        bloq_r = bloq_r.merge(dias_por_bloque_r, on='HR_INICIO_AUX_STR', how='left')
        bloq_r['N_DIAS'] = bloq_r['N_DIAS'].fillna(1).astype(int)
        bloq_r['Ocup_Pct'] = ((bloq_r['INSCRITOS'] / (total_capacidad * bloq_r['N_DIAS'])) * 100).round(2)
        bloq_r = bloq_r.sort_values('Ocup_Pct')
        bloques_bajos_resumen = bloq_r[bloq_r['Ocup_Pct'] < 20].head(5)['HR_INICIO_AUX_STR'].tolist()
    
    # Salas bajo promedio (excluyendo Sin Capacidad)
    n_bajo_prom_total = 0
    for campus_val in promedio_campus:
        prom_c = promedio_campus[campus_val]
        sub = tabla_sala_resumen[tabla_sala_resumen['CAMPUS_AUX'] == campus_val]
        n_bajo_prom_total += len(sub[sub['Ocupacion_Pct'] < prom_c])
    
    # Construir texto
    parrafos_resumen = []
    
    # Solo campus donde CAMPUS == CAMPUS_AUX (campus de análisis)
    campus_analisis = sorted(set(df_resumen['CAMPUS_AUX'].unique()) & set(promedio_campus.keys()))
    
    for campus_val in campus_analisis:
        prom_c = promedio_campus[campus_val]
        sub_campus = tabla_sala_resumen[tabla_sala_resumen['CAMPUS_AUX'] == campus_val]
        n_salas_campus = len(sub_campus)
        n_criticas_campus = len(sub_campus[sub_campus['Ocupacion_Pct'] <= 20])
        n_bajo_campus = len(sub_campus[sub_campus['Ocupacion_Pct'] < prom_c])
        
        texto = (f"El campus <b>{campus_val}</b> presenta una ocupación promedio de <b>{prom_c:.2f}%</b> "
                 f"sobre un total de {n_salas_campus} salas analizadas. ")
        
        if prom_c < 30:
            texto += "Este nivel indica una subutilización significativa de la infraestructura disponible. "
        elif prom_c < 60:
            texto += "Existen oportunidades de mejora en la asignación de espacios. "
        else:
            texto += "El nivel de uso se encuentra en un rango aceptable. "
        
        if n_criticas_campus > 0:
            texto += (f"Se identifican <b>{n_criticas_campus} salas en situación crítica</b> (ocupación igual o "
                      f"menor al 20%), las cuales requieren atención inmediata para evaluar su reasignación "
                      f"o reconversión. ")
        
        if n_bajo_campus > 0:
            texto += (f"Adicionalmente, {n_bajo_campus} salas se encuentran bajo el promedio del campus. ")
        
        parrafos_resumen.append(texto)
    
    # Info de bloques bajos
    if bloques_bajos_resumen:
        bloques_str = ', '.join(bloques_bajos_resumen)
        parrafos_resumen.append(
            f"En cuanto a la distribución horaria, se identifican los bloques <b>{bloques_str}</b> con baja "
            f"frecuencia de uso (bajo el 20% de ocupación), lo que representa una oportunidad para "
            f"optimizar la programación académica o destinar esos horarios a actividades complementarias."
        )
    
    # Conclusión general
    ocup_general = sum(promedio_campus.values()) / max(len(promedio_campus), 1)
    parrafos_resumen.append(
        f"La ocupación promedio general de los campus analizados es de <b>{ocup_general:.2f}%</b> "
        f"para el año <b>{anio}</b>. Se recomienda revisar las secciones siguientes para el detalle "
        f"de salas, bloques horarios y recomendaciones específicas."
    )
    
    for p in parrafos_resumen:
        story.append(Paragraph(p, style_parrafo_resumen))
    
    story.append(Spacer(1, 10))
    
    # =============================================
    # 2. FILTROS USADOS
    # =============================================
    story.append(Paragraph("2. Filtros Aplicados", style_subtitulo))
    
    filtros_data = [['Parámetro', 'Valor']]
    filtros_data.append(['AÑO', str(filtros_dict.get('anio', '-'))])
    filtros_data.append(['Campus', ', '.join([str(c) for c in filtros_dict.get('campus_list', [])])])
    filtros_data.append(['Edificios', ', '.join([str(e) for e in filtros_dict.get('campus_edificio_list', [])])])
    
    periodos = filtros_dict.get('periodo_list', [])
    filtros_data.append(['Período Académico', ', '.join([str(p) for p in periodos]) if periodos else 'Todos'])
    
    carreras = filtros_dict.get('desc_carrera_materia_list', [])
    carr_txt = ', '.join([str(c) for c in carreras[:5]]) + ('...' if len(carreras) > 5 else '') if carreras else 'Todas'
    filtros_data.append(['Carrera/Materia', carr_txt])
    
    fecha_ini = filtros_dict.get('fecha_inicio')
    fecha_f = filtros_dict.get('fecha_fin')
    filtros_data.append(['Rango Fechas', f'{fecha_ini} a {fecha_f}' if fecha_ini and fecha_f else 'Sin filtro'])
    
    excl_cap_f = filtros_dict.get('clasificacion_excluir_capacidad', [])
    filtros_data.append(['Clasif. Excluidas (Cap.)', ', '.join([str(c) for c in excl_cap_f]) if excl_cap_f else 'Ninguna'])
    
    excl_ocu_f = filtros_dict.get('clasificacion_excluir_ocupacion', [])
    filtros_data.append(['Clasif. Excluidas (Prog.)', ', '.join([str(c) for c in excl_ocu_f]) if excl_ocu_f else 'Ninguna'])
    
    if 'dias_sesion_list' in filtros_dict and filtros_dict['dias_sesion_list']:
        filtros_data.append(['Dias Sesion', ', '.join([str(d) for d in filtros_dict['dias_sesion_list']])])
    if 'bloque_inicio' in filtros_dict and filtros_dict['bloque_inicio']:
        filtros_data.append(['Bloque Horario', f"{filtros_dict['bloque_inicio']} - {filtros_dict.get('bloque_fin', '')}"])
    
    filtros_data.append(['Capacidad Total Campus', f"{total_capacidad:,.0f}"])
    
    t_filtros = Table(filtros_data, colWidths=[150, 340])
    t_filtros.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#1a237e')),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, -1), 8),
        ('BACKGROUND', (0, 1), (0, -1), colors.HexColor('#e8eaf6')),
        ('FONTNAME', (0, 1), (0, -1), 'Helvetica-Bold'),
        ('GRID', (0, 0), (-1, -1), 0.5, colors.HexColor('#bdbdbd')),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('LEFTPADDING', (0, 0), (-1, -1), 6),
        ('RIGHTPADDING', (0, 0), (-1, -1), 6),
        ('TOPPADDING', (0, 0), (-1, -1), 3),
        ('BOTTOMPADDING', (0, 0), (-1, -1), 3),
    ]))
    story.append(t_filtros)
    story.append(Spacer(1, 12))
    
    # =============================================
    # 3. TABLA DE % OCUPACIÓN - TODAS LAS SALAS
    # =============================================
    story.append(Paragraph("3. Porcentaje de Ocupación por Campus, Edificio, Clasificación, Rango y Sala", style_subtitulo))
    
    has_rango = 'Rango_Capacidad' in tabla_sala.columns
    
    # Encabezado
    if has_rango:
        tabla_data = [['Campus', 'Edificio', 'Clasificación', 'Rango Cap.', 'Sala', '% Ocup.']]
    else:
        tabla_data = [['Campus', 'Edificio', 'Clasificación', 'Sala', '% Ocup.']]
    
    for _, row in tabla_sala.iterrows():
        if has_rango:
            tabla_data.append([
                str(row['CAMPUS_AUX']), str(row['EDIFICIO']), str(row['CLASIFICACIÓN']),
                str(row.get('Rango_Capacidad', '')), str(row['SALA']),
                f"{row['Ocupacion_Pct']:.2f}%"
            ])
        else:
            tabla_data.append([
                str(row['CAMPUS_AUX']), str(row['EDIFICIO']), str(row['CLASIFICACIÓN']),
                str(row['SALA']), f"{row['Ocupacion_Pct']:.2f}%"
            ])
    
    if has_rango:
        col_widths_tabla = [45, 50, 70, 65, 50, 50]
    else:
        col_widths_tabla = [50, 60, 80, 55, 55]
    
    t_ocup = Table(tabla_data, colWidths=col_widths_tabla, repeatRows=1)
    
    estilo_tabla = [
        ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#1a237e')),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, -1), 6.5),
        ('GRID', (0, 0), (-1, -1), 0.3, colors.HexColor('#cccccc')),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('ALIGN', (-1, 0), (-1, -1), 'CENTER'),
        ('LEFTPADDING', (0, 0), (-1, -1), 3),
        ('RIGHTPADDING', (0, 0), (-1, -1), 3),
        ('TOPPADDING', (0, 0), (-1, -1), 2),
        ('BOTTOMPADDING', (0, 0), (-1, -1), 2),
    ]
    
    # Colores: ROJO <= 20%, AMARILLO < 60%, VERDE >= 60%
    for i in range(1, len(tabla_data)):
        try:
            val = float(tabla_data[i][-1].replace('%', '').replace(',', '.'))
            if val <= 20:
                estilo_tabla.append(('BACKGROUND', (0, i), (-1, i), colors.HexColor('#ffcdd2')))
            elif val < 60:
                estilo_tabla.append(('BACKGROUND', (0, i), (-1, i), colors.HexColor('#fff9c4')))
            else:
                estilo_tabla.append(('BACKGROUND', (0, i), (-1, i), colors.HexColor('#e8f5e9')))
        except:
            pass
    
    t_ocup.setStyle(TableStyle(estilo_tabla))
    story.append(t_ocup)
    story.append(Spacer(1, 4))
    
    # Leyenda
    leyenda = Table([['Rojo: <= 20%', 'Amarillo: 20% - 59%', 'Verde: >= 60%']], colWidths=[130, 130, 130])
    leyenda.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (0, 0), colors.HexColor('#ffcdd2')),
        ('BACKGROUND', (1, 0), (1, 0), colors.HexColor('#fff9c4')),
        ('BACKGROUND', (2, 0), (2, 0), colors.HexColor('#e8f5e9')),
        ('FONTSIZE', (0, 0), (-1, -1), 7),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('BOX', (0, 0), (-1, -1), 0.5, colors.grey),
    ]))
    story.append(leyenda)
    story.append(Spacer(1, 12))
    
    # =============================================
    # 4. SALAS BAJO EL PROMEDIO DE SU CAMPUS
    # =============================================
    story.append(Paragraph("4. Salas Bajo el Promedio de su Campus", style_subtitulo))
    
    salas_bajo_prom = []
    for _, row in tabla_sala.iterrows():
        campus = row['CAMPUS_AUX']
        prom_campus_val = promedio_campus.get(campus, 0)
        if row['Ocupacion_Pct'] < prom_campus_val:
            entry = {
                'Campus': campus, 'Edificio': row['EDIFICIO'],
                'Clasificación': row['CLASIFICACIÓN'], 'Sala': row['SALA'],
                'Ocupación': row['Ocupacion_Pct'], 'Promedio_Campus': prom_campus_val
            }
            if has_rango:
                entry['Rango_Capacidad'] = row.get('Rango_Capacidad', '')
            salas_bajo_prom.append(entry)
    
    df_bajo_prom = pd.DataFrame(salas_bajo_prom)
    
    # Excluir salas con Rango_Capacidad = "Sin Capacidad"
    if not df_bajo_prom.empty and 'Rango_Capacidad' in df_bajo_prom.columns:
        df_bajo_prom = df_bajo_prom[df_bajo_prom['Rango_Capacidad'].astype(str) != 'Sin Capacidad']
    
    if not df_bajo_prom.empty:
        for campus_val in sorted(df_bajo_prom['Campus'].unique()):
            sub = df_bajo_prom[df_bajo_prom['Campus'] == campus_val]
            prom_c = promedio_campus.get(campus_val, 0)
            story.append(Paragraph(
                f"Campus {campus_val} (Promedio: {prom_c:.2f}%) - {len(sub)} salas bajo promedio",
                style_seccion))
            
            if has_rango:
                bajo_data = [['Edificio', 'Clasificación', 'Rango Cap.', 'Sala', '% Ocup.', 'Diferencia']]
            else:
                bajo_data = [['Edificio', 'Clasificación', 'Sala', '% Ocup.', 'Diferencia']]
            
            for _, r in sub.sort_values('Ocupación').iterrows():
                diff = r['Ocupación'] - r['Promedio_Campus']
                if has_rango:
                    bajo_data.append([
                        str(r['Edificio']), str(r['Clasificación']), str(r.get('Rango_Capacidad', '')),
                        str(r['Sala']), f"{r['Ocupación']:.2f}%", f"{diff:.2f}%"
                    ])
                else:
                    bajo_data.append([
                        str(r['Edificio']), str(r['Clasificación']),
                        str(r['Sala']), f"{r['Ocupación']:.2f}%", f"{diff:.2f}%"
                    ])
            
            if has_rango:
                cw_bajo = [55, 70, 60, 50, 50, 50]
            else:
                cw_bajo = [70, 90, 60, 60, 65]
            
            t_bajo = Table(bajo_data, colWidths=cw_bajo, repeatRows=1)
            t_bajo.setStyle(TableStyle([
                ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#e65100')),
                ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
                ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                ('FONTSIZE', (0, 0), (-1, -1), 6.5),
                ('GRID', (0, 0), (-1, -1), 0.3, colors.HexColor('#cccccc')),
                ('BACKGROUND', (0, 1), (-1, -1), colors.HexColor('#fff3e0')),
                ('ALIGN', (-2, 0), (-1, -1), 'CENTER'),
                ('TOPPADDING', (0, 0), (-1, -1), 2),
                ('BOTTOMPADDING', (0, 0), (-1, -1), 2),
                ('LEFTPADDING', (0, 0), (-1, -1), 3),
                ('RIGHTPADDING', (0, 0), (-1, -1), 3),
            ]))
            story.append(t_bajo)
            story.append(Spacer(1, 6))
    else:
        story.append(Paragraph("No se encontraron salas bajo el promedio de su campus.", style_normal))
    
    story.append(Spacer(1, 8))
    
    # =============================================
    # 5. SALAS CRÍTICAS (OCUPACIÓN <= 20%)
    # =============================================
    story.append(Paragraph("5. Salas Críticas (Ocupación Igual o Menor al 20%)", style_subtitulo))
    
    salas_criticas = tabla_sala[tabla_sala['Ocupacion_Pct'] <= 20].copy()
    
    # Excluir salas con Rango_Capacidad = "Sin Capacidad"
    if 'Rango_Capacidad' in salas_criticas.columns:
        salas_criticas = salas_criticas[salas_criticas['Rango_Capacidad'].astype(str) != 'Sin Capacidad']
    
    if not salas_criticas.empty:
        story.append(Paragraph(
            f"Se identificaron {len(salas_criticas)} salas con ocupación crítica (igual o menor al 20%).",
            style_alerta))
        story.append(Spacer(1, 4))
        
        if has_rango:
            crit_data = [['Campus', 'Edificio', 'Clasificación', 'Rango Cap.', 'Sala', '% Ocup.']]
        else:
            crit_data = [['Campus', 'Edificio', 'Clasificación', 'Sala', '% Ocup.']]
        
        for _, r in salas_criticas.sort_values('Ocupacion_Pct').iterrows():
            if has_rango:
                crit_data.append([
                    str(r['CAMPUS_AUX']), str(r['EDIFICIO']), str(r['CLASIFICACIÓN']),
                    str(r.get('Rango_Capacidad', '')), str(r['SALA']), f"{r['Ocupacion_Pct']:.2f}%"
                ])
            else:
                crit_data.append([
                    str(r['CAMPUS_AUX']), str(r['EDIFICIO']), str(r['CLASIFICACIÓN']),
                    str(r['SALA']), f"{r['Ocupacion_Pct']:.2f}%"
                ])
        
        if has_rango:
            cw_crit = [45, 50, 70, 65, 50, 50]
        else:
            cw_crit = [50, 60, 80, 55, 55]
        
        t_crit = Table(crit_data, colWidths=cw_crit, repeatRows=1)
        t_crit.setStyle(TableStyle([
            ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#b71c1c')),
            ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
            ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
            ('FONTSIZE', (0, 0), (-1, -1), 6.5),
            ('GRID', (0, 0), (-1, -1), 0.3, colors.HexColor('#cccccc')),
            ('BACKGROUND', (0, 1), (-1, -1), colors.HexColor('#ffebee')),
            ('ALIGN', (-1, 0), (-1, -1), 'CENTER'),
            ('TOPPADDING', (0, 0), (-1, -1), 2),
            ('BOTTOMPADDING', (0, 0), (-1, -1), 2),
            ('LEFTPADDING', (0, 0), (-1, -1), 3),
            ('RIGHTPADDING', (0, 0), (-1, -1), 3),
        ]))
        story.append(t_crit)
    else:
        story.append(Paragraph("No se encontraron salas con ocupación igual o menor al 20%.", style_normal))
    
    story.append(Spacer(1, 12))
    
    # =============================================
    # 6. BLOQUES HORARIOS DE MENOR OCUPACIÓN
    # =============================================
    story.append(Paragraph("6. Bloques Horarios de Menor Ocupación", style_subtitulo))
    story.append(Paragraph(
        "<i>Nota: Se excluyen los días sábado y domingo del cálculo de bloques. "
        "El % de ocupación se calcula como SUM(Inscritos) / (Capacidad Total Campus x Cantidad de Dias L-V del bloque). "
        "El flujo de estudiantes se calcula como SUM(Inscritos) / Capacidad Total Campus (solo días lunes a viernes).</i>",
        style_small))
    story.append(Spacer(1, 4))
    
    if 'HR_INICIO_AUX_STR' in df_work.columns and 'DIA_SESION' in df_work.columns:
        # Excluir sábado y domingo
        df_bloques = df_work[~df_work['DIA_SESION'].isin(['SABADO', 'DOMINGO'])].copy()
        
        if not df_bloques.empty:
            # Sumatoria de inscritos por bloque
            bloq_inscritos = df_bloques.groupby('HR_INICIO_AUX_STR')['INSCRITOS'].sum().reset_index()
            bloq_inscritos.columns = ['HR_INICIO_AUX_STR', 'SUM_INSCRITOS']
            
            # Contar días únicos L-V por bloque
            dias_por_bloque = df_bloques.groupby('HR_INICIO_AUX_STR')['DIA_SESION'].nunique().reset_index()
            dias_por_bloque.columns = ['HR_INICIO_AUX_STR', 'N_DIAS']
            
            bloq_data_calc = bloq_inscritos.merge(dias_por_bloque, on='HR_INICIO_AUX_STR', how='left')
            bloq_data_calc['N_DIAS'] = bloq_data_calc['N_DIAS'].fillna(1).astype(int)
            
            # % Ocupación = SUM(INSCRITOS) / (total_capacidad * N_DIAS) * 100
            bloq_data_calc['Ocup_Pct'] = ((bloq_data_calc['SUM_INSCRITOS'] / (total_capacidad * bloq_data_calc['N_DIAS'])) * 100).round(2)
            
            # Flujo estudiantes (solo L-V) = SUM(INSCRITOS L-V) / total_capacidad
            df_bloques_lv = df_work[~df_work['DIA_SESION'].isin(['SABADO', 'DOMINGO'])].copy()
            flujo_lv = df_bloques_lv.groupby('HR_INICIO_AUX_STR')['INSCRITOS'].sum().reset_index()
            flujo_lv.columns = ['HR_INICIO_AUX_STR', 'SUM_INSCRITOS_LV']
            flujo_lv['Flujo_Estudiantes'] = (flujo_lv['SUM_INSCRITOS_LV'] / total_capacidad).round(2)
            
            bloq_data_calc = bloq_data_calc.merge(flujo_lv[['HR_INICIO_AUX_STR', 'Flujo_Estudiantes']], 
                                                    on='HR_INICIO_AUX_STR', how='left')
            bloq_data_calc = bloq_data_calc.sort_values('Ocup_Pct').reset_index(drop=True)
            
            bloq_tabla = [['Bloque Horario', '% Ocupación', 'Flujo Estudiantes']]
            for _, r in bloq_data_calc.iterrows():
                bloq_tabla.append([
                    str(r['HR_INICIO_AUX_STR']),
                    f"{r['Ocup_Pct']:.2f}%",
                    f"{r['Flujo_Estudiantes']:.2f}"
                ])
            
            t_bloq = Table(bloq_tabla, colWidths=[120, 110, 110], repeatRows=1)
            t_bloq.setStyle(TableStyle([
                ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#1a237e')),
                ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
                ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                ('FONTSIZE', (0, 0), (-1, -1), 8),
                ('GRID', (0, 0), (-1, -1), 0.4, colors.HexColor('#cccccc')),
                ('ALIGN', (1, 0), (2, -1), 'CENTER'),
                ('TOPPADDING', (0, 0), (-1, -1), 3),
                ('BOTTOMPADDING', (0, 0), (-1, -1), 3),
            ]))
            story.append(t_bloq)
            story.append(Spacer(1, 8))
            
            # Ocupación promedio por día (excluyendo sab y dom, misma lógica)
            story.append(Paragraph("Ocupación promedio por día (Lunes a Viernes):", style_seccion))
            story.append(Paragraph(
                "<i>Nota: Se excluyen sábado y domingo. Flujo de estudiantes calculado solo lunes a viernes.</i>",
                style_small))
            story.append(Spacer(1, 4))
            
            dias_inscritos = df_bloques_lv.groupby('DIA_SESION')['INSCRITOS'].sum().reset_index()
            dias_inscritos.columns = ['DIA_SESION', 'SUM_INSCRITOS']
            
            # Para % ocupación por día: contar bloques únicos por día
            bloques_por_dia = df_bloques_lv.groupby('DIA_SESION')['HR_INICIO_AUX_STR'].nunique().reset_index()
            bloques_por_dia.columns = ['DIA_SESION', 'N_BLOQUES']
            
            dias_calc = dias_inscritos.merge(bloques_por_dia, on='DIA_SESION', how='left')
            dias_calc['N_BLOQUES'] = dias_calc['N_BLOQUES'].fillna(1).astype(int)
            dias_calc['Ocup_Pct'] = ((dias_calc['SUM_INSCRITOS'] / (total_capacidad * dias_calc['N_BLOQUES'])) * 100).round(2)
            dias_calc['Flujo_Estudiantes'] = (dias_calc['SUM_INSCRITOS'] / total_capacidad).round(2)
            
            orden_dias = ['LUNES', 'MARTES', 'MIERCOLES', 'JUEVES', 'VIERNES']
            dias_calc['orden'] = dias_calc['DIA_SESION'].apply(lambda x: orden_dias.index(x) if x in orden_dias else 99)
            dias_calc = dias_calc.sort_values('orden')
            
            dias_tabla = [['Dia', '% Ocupación', 'Flujo Estudiantes']]
            for _, r in dias_calc.iterrows():
                dias_tabla.append([str(r['DIA_SESION']), f"{r['Ocup_Pct']:.2f}%", f"{r['Flujo_Estudiantes']:.2f}"])
            
            t_dias = Table(dias_tabla, colWidths=[120, 110, 110], repeatRows=1)
            t_dias.setStyle(TableStyle([
                ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#37474f')),
                ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
                ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                ('FONTSIZE', (0, 0), (-1, -1), 8),
                ('GRID', (0, 0), (-1, -1), 0.4, colors.HexColor('#cccccc')),
                ('ALIGN', (1, 0), (2, -1), 'CENTER'),
                ('TOPPADDING', (0, 0), (-1, -1), 3),
                ('BOTTOMPADDING', (0, 0), (-1, -1), 3),
            ]))
            story.append(t_dias)
        else:
            story.append(Paragraph("No hay datos de bloques horarios para días lunes a viernes.", style_normal))
    else:
        story.append(Paragraph("No se dispone de datos de bloques horarios.", style_normal))
    
    story.append(Spacer(1, 12))
    
    # =============================================
    # 7. RECOMENDACIONES (DINÁMICAS)
    # =============================================
    story.append(Paragraph("7. Recomendaciones", style_subtitulo))
    
    recomendaciones = []
    
    # R1: Salas críticas <= 20%
    # salas_criticas ya excluye "Sin Capacidad" desde el paso anterior
    n_criticas_final = len(salas_criticas) if not salas_criticas.empty else 0
    if n_criticas_final > 0:
        salas_top5 = salas_criticas.sort_values('Ocupacion_Pct').head(5)
        salas_nombres = ', '.join([
            f"Sala {r['SALA']} ({r['CAMPUS_AUX']}-{r['EDIFICIO']}, {r['Ocupacion_Pct']:.1f}%)" 
            for _, r in salas_top5.iterrows()])
        recomendaciones.append(
            f"CRÍTICO: Se identificaron {n_criticas_final} salas con ocupación igual o menor al 20%. "
            f"Se recomienda evaluar la reasignación o reconversión de las más críticas: {salas_nombres}. "
            f"Considerar su uso para actividades complementarias, estudios libres o eventos."
        )
    
    # (Recomendación de ocupación promedio general eliminada por solicitud)
    
    # R3: Bloques horarios bajos (excluyendo sab/dom)
    if 'HR_INICIO_AUX_STR' in df_work.columns and 'DIA_SESION' in df_work.columns:
        df_blq_rec = df_work[~df_work['DIA_SESION'].isin(['SABADO', 'DOMINGO'])].copy()
        if not df_blq_rec.empty:
            blq_rec = df_blq_rec.groupby('HR_INICIO_AUX_STR')['INSCRITOS'].sum().reset_index()
            dias_blq_rec = df_blq_rec.groupby('HR_INICIO_AUX_STR')['DIA_SESION'].nunique().reset_index()
            dias_blq_rec.columns = ['HR_INICIO_AUX_STR', 'N_DIAS']
            blq_rec = blq_rec.merge(dias_blq_rec, on='HR_INICIO_AUX_STR', how='left')
            blq_rec['Pct'] = ((blq_rec['INSCRITOS'] / (total_capacidad * blq_rec['N_DIAS'])) * 100).round(2)
            bloques_muy_bajos = blq_rec[blq_rec['Pct'] <= 20]
            if not bloques_muy_bajos.empty:
                horas = ', '.join(bloques_muy_bajos['HR_INICIO_AUX_STR'].tolist())
                recomendaciones.append(
                    f"Los bloques horarios {horas} presentan ocupación igual o menor al 20% (excluyendo "
                    f"sábado y domingo). Se recomienda evaluar si es viable reducir la oferta en estos "
                    f"horarios o incentivar su uso con actividades académicas complementarias."
                )
    
    # R4: Salas bajo promedio
    n_bajo_prom_f = len(df_bajo_prom) if not df_bajo_prom.empty else 0
    total_salas_f = len(tabla_sala)
    if total_salas_f > 0:
        pct_bajo = (n_bajo_prom_f / total_salas_f) * 100
        if pct_bajo > 40:
            recomendaciones.append(
                f"El {pct_bajo:.1f}% de las salas ({n_bajo_prom_f} de {total_salas_f}) estan bajo el "
                f"promedio de su campus. Esto sugiere una distribución desigual y concentración de uso "
                f"en pocas salas."
            )
    
    # R5: Clasificaciones con baja ocupación
    if 'CLASIFICACIÓN' in df_work.columns:
        clasif_ocup = df_work.groupby('CLASIFICACIÓN')[col_ocup].mean().reset_index()
        clasif_ocup['Pct'] = (clasif_ocup[col_ocup] * 100).round(2)
        clasif_bajas = clasif_ocup[clasif_ocup['Pct'] < 15]
        if not clasif_bajas.empty:
            tipos = ', '.join([f"{r['CLASIFICACIÓN']} ({r['Pct']:.1f}%)" for _, r in clasif_bajas.iterrows()])
            recomendaciones.append(
                f"Las siguientes clasificaciones de sala tienen ocupación promedio bajo 15%: {tipos}. "
                f"Se recomienda evaluar si estos tipos de espacio se ajustan a la demanda actual."
            )
    
    # R6: Campus cruzados (CAMPUS != CAMPUS_AUX)
    if 'CAMPUS' in df_work.columns and 'CAMPUS_AUX' in df_work.columns:
        df_cruzados = df_work[df_work['CAMPUS'] != df_work['CAMPUS_AUX']].copy()
        if not df_cruzados.empty:
            n_salas_cruzadas = df_cruzados['SALA'].nunique()
            recomendaciones.append(
                f"ALERTA: Se detectaron {n_salas_cruzadas} salas programadas en un campus diferente al "
                f"campus físico donde se encuentran (CAMPUS de programación difiere de CAMPUS_AUX). "
                f"Se recomienda revisar estas asignaciones para asegurar consistencia entre la "
                f"programación y la ubicación real de los espacios."
            )
    
    # R7: Salas sin capacidad y salas "Sin Clasificación" - buscar en ambas fuentes
    n_sin_cap = 0
    n_sin_clasif = 0
    
    # Buscar desde df3 (planta física)
    if 'Capacidad Máxima Sala' in df3_ref.columns:
        filtro_df3_rec = df3_ref['Campus Edificio'].isin(campus_edificio_list)
        if excl_cap:
            filtro_df3_rec = filtro_df3_rec & (~df3_ref['CLASIFICACIÓN'].isin(excl_cap))
        df3_check = df3_ref[filtro_df3_rec].copy()
        
        n_sin_cap = len(df3_check[
            (df3_check['Capacidad Máxima Sala'].isna()) | (df3_check['Capacidad Máxima Sala'] == 0)
        ])
        
        n_sin_clasif = len(df3_check[
            (df3_check['CLASIFICACIÓN'].isna()) | 
            (df3_check['CLASIFICACIÓN'].astype(str).str.strip() == '') |
            (df3_check['CLASIFICACIÓN'].astype(str).str.strip() == 'Sin Clasificación') |
            (df3_check['CLASIFICACIÓN'].astype(str).str.strip() == 'Sin Clasificacion') |
            (df3_check['CLASIFICACIÓN'].astype(str) == 'nan')
        ])
    
    # Complementar con datos filtrados (Rango_Capacidad = "Sin Capacidad")
    if 'Rango_Capacidad' in df_work.columns:
        salas_sin_cap_work = df_work[df_work['Rango_Capacidad'].astype(str) == 'Sin Capacidad']['SALA'].nunique()
        if salas_sin_cap_work > n_sin_cap:
            n_sin_cap = salas_sin_cap_work
    
    if 'CLASIFICACIÓN' in df_work.columns:
        clasif_vals = ['Sin Clasificación', 'Sin Clasificacion', 'nan', '']
        salas_sin_clasif_work = df_work[
            df_work['CLASIFICACIÓN'].astype(str).str.strip().isin(clasif_vals)
        ]['SALA'].nunique()
        if salas_sin_clasif_work > n_sin_clasif:
            n_sin_clasif = salas_sin_clasif_work
    
    # Siempre agregar esta recomendación (incluso si son 0, para que se vea la verificación)
    texto_datos = "RECTIFICACIÓN DE DATOS: "
    partes = []
    if n_sin_cap > 0:
        partes.append(f"{n_sin_cap} salas no tienen capacidad registrada o tienen capacidad 0")
    if n_sin_clasif > 0:
        partes.append(f"{n_sin_clasif} salas figuran con 'Sin Clasificación' en su clasificación")
    
    if partes:
        texto_datos += ' y '.join(partes) + ". "
        texto_datos += ("Se sugiere rectificar estos datos por el área correspondiente en el sistema "
                       "de planta física, ya que afectan la precisión de los análisis de ocupación.")
        recomendaciones.append(texto_datos)
    
    # Escribir recomendaciones con ancho completo
    if recomendaciones:
        page_width = letter[0] - 3*cm  # ancho disponible
        for i, rec in enumerate(recomendaciones, 1):
            # Usar Paragraph dentro de Table para que el texto haga wrap correctamente
            para = Paragraph(f"<b>{i}.</b> {rec}", 
                           ParagraphStyle('RecInner', parent=styles['Normal'], fontSize=8, leading=11))
            rec_table = Table([[para]], colWidths=[page_width])
            
            bg_color = colors.HexColor('#ffebee') if ('CRÍTICO' in rec or 'ALERTA' in rec or 'DATOS' in rec) else colors.HexColor('#e3f2fd')
            rec_table.setStyle(TableStyle([
                ('BACKGROUND', (0, 0), (-1, -1), bg_color),
                ('BOX', (0, 0), (-1, -1), 0.5, colors.HexColor('#90a4ae')),
                ('LEFTPADDING', (0, 0), (-1, -1), 8),
                ('RIGHTPADDING', (0, 0), (-1, -1), 8),
                ('TOPPADDING', (0, 0), (-1, -1), 5),
                ('BOTTOMPADDING', (0, 0), (-1, -1), 5),
            ]))
            story.append(rec_table)
            story.append(Spacer(1, 4))
    else:
        story.append(Paragraph("No se generaron recomendaciones específicas.", style_normal))
    
    # =============================================
    # PIE DE PÁGINA
    # =============================================
    story.append(Spacer(1, 20))
    story.append(HRFlowable(width="100%", thickness=1, color=colors.HexColor('#bdbdbd')))
    story.append(Paragraph(
        f"Documento generado automáticamente | {fecha_gen} | Total registros analizados: {len(df_filtrado):,}",
        ParagraphStyle('Pie', parent=styles['Normal'], fontSize=7,
                       alignment=TA_CENTER, textColor=colors.HexColor('#999999'))))
    
    # =============================================
    # GENERAR PDF
    # =============================================
    try:
        doc.build(story)
        n_bajo_prom_f = len(df_bajo_prom) if not df_bajo_prom.empty else 0
        print(f"✅ Resumen Ejecutivo generado: {filename}")
        print(f"   📄 Ubicacion: {os.path.abspath(filename)}")
        print(f"   📊 Salas analizadas: {len(tabla_sala)}")
        print(f"   ⚠️ Salas críticas (<=20%): {n_criticas_final}")
        print(f"   📉 Salas bajo promedio campus: {n_bajo_prom_f}")
        print(f"   💡 Recomendaciones generadas: {len(recomendaciones)}")
    except Exception as e:
        print(f"❌ Error al generar PDF: {str(e)}")
        import traceback
        traceback.print_exc()

print("✅ Función generar_resumen_ejecutivo_pdf V2 cargada correctamente.")
```

```python
######### FUNCIÓN PRINCIPAL PARA GENERAR HEATMAP CON FILTROS #####################

def generar_heatmap_interactivo(campus_edificio_list, clasificacion_excluir_capacidad, clasificacion_excluir_ocupacion, 
                                anio, campus_list, periodo_list, desc_carrera_materia_list, fecha_inicio, fecha_fin):
    """
    Genera el heatmap con los filtros seleccionados
    
    Parámetros:
    - campus_edificio_list: list - Lista de edificios del campus a filtrar
    - clasificacion_excluir_capacidad: list - Lista de clasificaciones a excluir en df3 (capacidad)
    - clasificacion_excluir_ocupacion: list - Lista de clasificaciones a excluir en df_ocupacion
    - anio: int - Año a filtrar
    - campus_list: list - Lista de campus a filtrar
    - periodo_list: list - Lista de periodos académicos a filtrar
    - desc_carrera_materia_list: list - Lista de descripciones de carrera/materia a filtrar
    - fecha_inicio: date - Fecha de inicio del rango
    - fecha_fin: date - Fecha de fin del rango
    """
    
    ## 1.- GENERACIÓN Y FILTRO PARA GENERAR MATRIZ DE CAPACIDAD
    
    # Crear filtro dinámico para df3 (ahora con múltiples edificios)
    filtro_df3 = df3['Campus Edificio'].isin(campus_edificio_list)
    
    # Aplicar filtro de clasificación (excluir las seleccionadas) - CAPACIDAD
    if clasificacion_excluir_capacidad:
        filtro_df3 = filtro_df3 & (~df3['CLASIFICACIÓN'].isin(clasificacion_excluir_capacidad))
    
    # Calcular total de capacidad con filtros aplicados
    total_capacidad = df3.loc[filtro_df3, 'Capacidad Máxima Sala'].sum()
    
    if total_capacidad == 0:
        print("⚠️ No hay datos de capacidad con los filtros seleccionados")
        return
    
    ## 2.- POBLAMIENTO DE BLOQUES HORARIO CON ESPACIO FISICO
    df2_capacidad = df2.replace('X', total_capacidad)
    
    ## 3.- GENERACIÓN DE MATRIZ DE OCUPACIÓN
 
 ##################################################################
    # Filtrar por REG_UNICO_LLAV2 = 1 y DIA_SESION no vacío
    df_ocupacion = df_aperturado[
        (df_aperturado['REG_UNICO_LLAV2'] == 1) & 
        (df_aperturado['DIA_SESION'].notna()) &
        (df_aperturado['DIA_SESION'] != '') &
        (df_aperturado['SALA'] != '')
    ]
 ##################################################################   
    # Aplicar filtros dinámicos de AÑO y CAMPUS (múltiples campus)
    df_ocupacion = df_ocupacion[
        (df_ocupacion['AÑO'] == anio) & 
        (df_ocupacion['CAMPUS'].isin(campus_list))
    ]
    
    # *** NUEVO: Aplicar filtro de PERIODO (múltiples periodos) ***
    if periodo_list:
        df_ocupacion = df_ocupacion[df_ocupacion['PERIODO'].isin(periodo_list)]
    
    # *** NUEVO: Aplicar filtro de DESCRIPCIÓN CARRERA_MATERIA (múltiples) ***
    if desc_carrera_materia_list:
        df_ocupacion = df_ocupacion[df_ocupacion['DESCRIPCIÓN CARRERA_MATERIA'].isin(desc_carrera_materia_list)]
    
     # *** NUEVO: Aplicar filtro de RANGO DE FECHAS ***
    if fecha_inicio and fecha_fin:
        # Asegurar que las columnas de fecha estén en formato datetime (desde números seriales de Excel)
        if 'FECHA_INI2' in df_ocupacion.columns and 'FECHA_TERM2' in df_ocupacion.columns:
            df_ocupacion['FECHA_INI2'] = pd.to_datetime(df_ocupacion['FECHA_INI2'], origin='1899-12-30', unit='D', errors='coerce')
            df_ocupacion['FECHA_TERM2'] = pd.to_datetime(df_ocupacion['FECHA_TERM2'], origin='1899-12-30', unit='D', errors='coerce')
            
            # Convertir fechas de filtro a datetime
            fecha_inicio_dt = pd.to_datetime(fecha_inicio)
            fecha_fin_dt = pd.to_datetime(fecha_fin)
            
            # Filtrar registros que se solapen con el rango seleccionado
            # (inicio del registro <= fin del rango) Y (fin del registro >= inicio del rango)
            df_ocupacion = df_ocupacion[
                (df_ocupacion['FECHA_INI2'] <= fecha_fin_dt) & 
                (df_ocupacion['FECHA_TERM2'] >= fecha_inicio_dt)
            ]
            
            # *** CREAR COLUMNAS AUXILIARES ***
            # FECHA_INI_AUX: usar fecha_inicio del filtro, EXCEPTO si FECHA_INI2 es mayor
            df_ocupacion['FECHA_INI_AUX'] = df_ocupacion['FECHA_INI2'].apply(
                lambda x: x if pd.notna(x) and x >= fecha_inicio_dt else fecha_inicio_dt
            )
            
            # FECHA_TERM_AUX: usar fecha_fin del filtro, EXCEPTO si FECHA_TERM2 es menor
            df_ocupacion['FECHA_TERM_AUX'] = df_ocupacion['FECHA_TERM2'].apply(
                lambda x: x if pd.notna(x) and x <= fecha_fin_dt else fecha_fin_dt
            )
    else:
        # Si no se usa filtro de fechas, copiar las columnas originales
        if 'FECHA_INI2' in df_ocupacion.columns and 'FECHA_TERM2' in df_ocupacion.columns:
            df_ocupacion['FECHA_INI2'] = pd.to_datetime(df_ocupacion['FECHA_INI2'], origin='1899-12-30', unit='D', errors='coerce')
            df_ocupacion['FECHA_TERM2'] = pd.to_datetime(df_ocupacion['FECHA_TERM2'], origin='1899-12-30', unit='D', errors='coerce')
            df_ocupacion['FECHA_INI_AUX'] = df_ocupacion['FECHA_INI2']
            df_ocupacion['FECHA_TERM_AUX'] = df_ocupacion['FECHA_TERM2']
    
    # Aplicar filtro de clasificación en df_ocupacion (INDEPENDIENTE)
    if clasificacion_excluir_ocupacion:
        df_ocupacion = df_ocupacion[~df_ocupacion['CLASIFICACIÓN'].isin(clasificacion_excluir_ocupacion)]
    
    if df_ocupacion.empty:
        print("⚠️ No hay datos de ocupación con los filtros seleccionados")
        return
    
    # Aplicar filtro de clasificación en df_ocupacion (INDEPENDIENTE)
    if clasificacion_excluir_ocupacion:
        df_ocupacion = df_ocupacion[~df_ocupacion['CLASIFICACIÓN'].isin(clasificacion_excluir_ocupacion)]
    
    if df_ocupacion.empty:
        print("⚠️ No hay datos de ocupación con los filtros seleccionados")
        return
    
    # *** GUARDAR DataFrame filtrado como variable global para exportación ***
    global df_ocupacion_filtrado
    df_ocupacion_filtrado = df_ocupacion.copy()
    
    # Definir el orden correcto de los días de la semana
    orden_dias = ['LUNES', 'MARTES', 'MIERCOLES', 'JUEVES', 'VIERNES', 'SABADO', 'DOMINGO']
    
    ## 4.- CREACIÓN DE MATRIZ DE PORCENTAJE
    
    # Asegurarse de que df_ocupacion tenga las horas como strings
    df_ocupacion['HR_INICIO_AUX_STR'] = df_ocupacion['HR_INICIO_AUX_STR'].astype(str)
    
    # Crear matriz pivote de ocupación
    matriz_ocupacion = pd.pivot_table(
        df_ocupacion,
        values='INSCRITOS',
        index='HR_INICIO_AUX_STR',
        columns='DIA_SESION',
        aggfunc='sum',
        fill_value=0
    )
    
    # Asegurar que todas las columnas de días existan
    for dia in orden_dias:
        if dia not in matriz_ocupacion.columns:
            matriz_ocupacion[dia] = 0
    
    # Ordenar columnas
    matriz_ocupacion = matriz_ocupacion[orden_dias]
    
    # Preparar df2_capacidad para el cruce
    df2_capacidad['HORA_INICIO_STR'] = df2_capacidad['HORA INICIO'].astype(str)
    
    # Crear DataFrame de capacidad en formato matriz
    df_capacidad_melted = df2_capacidad.melt(
        id_vars=['BLOQUE', 'HORA_INICIO_STR'],
        value_vars=orden_dias,
        var_name='DIA',
        value_name='CAPACIDAD_TOTAL'
    )
    
    # Preparar matriz_ocupacion para el merge
    matriz_ocupacion_reset = matriz_ocupacion.reset_index()
    matriz_ocupacion_reset = matriz_ocupacion_reset.rename(columns={'HR_INICIO_AUX_STR': 'HORA_INICIO_STR'})
    
    # Derretir la matriz de ocupación
    ocupacion_melted = matriz_ocupacion_reset.melt(
        id_vars=['HORA_INICIO_STR'],
        value_vars=orden_dias,
        var_name='DIA',
        value_name='OCUPACION'
    )
    
    # Fusionar ocupación con capacidad
    df_combinado = pd.merge(
        df_capacidad_melted,
        ocupacion_melted,
        on=['HORA_INICIO_STR', 'DIA'],
        how='left'
    )
    
    # Rellenar NaN en ocupación con 0
    df_combinado['OCUPACION'] = df_combinado['OCUPACION'].fillna(0)
    
    # Calcular porcentaje de ocupación
    df_combinado['PORCENTAJE'] = (df_combinado['OCUPACION'] / df_combinado['CAPACIDAD_TOTAL']) * 100
    
    # Formatear porcentaje
    df_combinado['PORCENTAJE_FORMATEADO'] = df_combinado['PORCENTAJE'].apply(
        lambda x: f"{x:.2f}".replace('.', ',') + '%' if not pd.isna(x) else "0,00%"
    )
    
    # Crear matriz pivote de porcentajes
    df_mapa_calor = df_combinado.pivot_table(
        index=['BLOQUE', 'HORA_INICIO_STR'],
        columns='DIA',
        values='PORCENTAJE_FORMATEADO',
        aggfunc='first'
    )
    
    # Reindexar para mantener orden de días
    df_mapa_calor = df_mapa_calor[orden_dias]
    
    # Resetear índice
    df_mapa_calor = df_mapa_calor.reset_index()
    
    # Añadir columnas de HORA INICIO y HORA FIN
    horas_dict = df2_capacidad.set_index('HORA_INICIO_STR')[['HORA INICIO', 'HORA FIN']].to_dict('index')
    
    df_mapa_calor['HORA INICIO'] = df_mapa_calor['HORA_INICIO_STR'].map(
        lambda x: horas_dict.get(x, {}).get('HORA INICIO', '')
    )
    df_mapa_calor['HORA FIN'] = df_mapa_calor['HORA_INICIO_STR'].map(
        lambda x: horas_dict.get(x, {}).get('HORA FIN', '')
    )
    
    # Reordenar columnas
    column_order = ['BLOQUE', 'HORA INICIO', 'HORA FIN'] + orden_dias
    df_mapa_calor = df_mapa_calor[column_order]
    
    # Ordenar por BLOQUE
    df_mapa_calor = df_mapa_calor.sort_values('BLOQUE').reset_index(drop=True)
    
    ## 5.- Convertir porcentajes a números
    def porcentaje_a_numero(porcentaje_str):
        if isinstance(porcentaje_str, str):
            try:
                return float(porcentaje_str.replace('%', '').replace(',', '.'))
            except:
                return 0.0
        return 0.0
    
    ## 6.- CREACIÓN DEL HEATMAP INTERACTIVO
    df_heatmap = df_mapa_calor.copy()
    
    # Crear matriz numérica de porcentajes
    matriz_numerica = df_heatmap[orden_dias].applymap(porcentaje_a_numero)
    bloque_labels = 'B' + df_heatmap['BLOQUE'].astype(str) + ' ' + \
                    df_heatmap['HORA INICIO'].astype(str) + '-' + \
                    df_heatmap['HORA FIN'].astype(str)
    
    # Crear matriz para plotly
    z = matriz_numerica.values
    x = orden_dias
    y = bloque_labels
    
    fig = go.Figure(data=go.Heatmap(
        z=z,
        x=x,
        y=y,
        colorscale='YlOrRd',
        zmin=0,
        zmax=100,
        text=[[f'{val:.2f}%' for val in row] for row in z],
        texttemplate='%{text}',
        textfont={"size": 10},
        hovertemplate='<b>Día:</b> %{x}<br>' +
                     '<b>Bloque:</b> %{y}<br>' +
                     '<b>Ocupación:</b> %{z:.2f}%<extra></extra>',
        colorbar=dict(
            title="Ocupación (%)",
            ticksuffix="%"
        )
    ))
    
    # Título dinámico con información de filtros
    campus_str = ', '.join(campus_list) if len(campus_list) <= 3 else f"{len(campus_list)} campus"
    edificio_str = ', '.join(campus_edificio_list) if len(campus_edificio_list) <= 3 else f"{len(campus_edificio_list)} edificios"
    
    titulo = f'Mapa de Calor - Ocupación Académica de Campus<br>' + \
             f'<sub>Campus: {campus_str} | Edificios: {edificio_str} | Año: {anio}</sub>'
    
    if fecha_inicio and fecha_fin:
        titulo += f'<br><sub>Periodo: {fecha_inicio} a {fecha_fin}</sub>'
    
    fig.update_layout(
        title=titulo,
        xaxis_title='Día de la Semana',
        yaxis_title='Bloque Horario',
        yaxis={'autorange': 'reversed'},
        width=1000,
        height=800,
        font=dict(size=12)
    )
    
    fig.show()
    
    # Mostrar información adicional
    print(f"\n📊 Resumen de datos:")
    print(f"   • Capacidad total: {total_capacidad:,.0f}")
    print(f"   • Registros de ocupación filtrados: {len(df_ocupacion):,}")
    print(f"   • Edificios seleccionados: {len(campus_edificio_list)}")
    print(f"   • Campus seleccionados: {len(campus_list)}")
    if periodo_list:
        print(f"   • Periodos seleccionados: {', '.join(map(str, periodo_list))}")
    if desc_carrera_materia_list:
        carreras_str = ', '.join(desc_carrera_materia_list[:3])
        if len(desc_carrera_materia_list) > 3:
            carreras_str += f" (y {len(desc_carrera_materia_list) - 3} más)"
        print(f"   • Carreras/Materias: {carreras_str}")
    if fecha_inicio and fecha_fin:
        print(f"   • Rango de fechas: {fecha_inicio} a {fecha_fin}")
    if clasificacion_excluir_capacidad:
        print(f"   • Clasificaciones excluidas (Capacidad): {', '.join(clasificacion_excluir_capacidad)}")
    if clasificacion_excluir_ocupacion:
        print(f"   • Clasificaciones excluidas (Ocupación): {', '.join(clasificacion_excluir_ocupacion)}")


######### CREAR WIDGETS DE FILTRO #####################

# Obtener valores únicos para los filtros (convertir a string para evitar errores de tipo)
campus_edificios = sorted(df3['Campus Edificio'].dropna().unique(), key=str)
clasificaciones_capacidad = sorted(df3['CLASIFICACIÓN'].dropna().unique(), key=str)
clasificaciones_ocupacion = sorted(df_aperturado['CLASIFICACIÓN'].dropna().unique(), key=str)
anios = sorted(df_aperturado['AÑO'].dropna().unique(), key=str)
campus_list_options = sorted(df_aperturado['CAMPUS'].dropna().unique(), key=str)
periodos = sorted(df_aperturado['PERIODO'].dropna().unique(), key=str)
desc_carreras_materias = sorted(df_aperturado['DESCRIPCIÓN CARRERA_MATERIA'].dropna().unique(), key=str)

# Obtener rango de fechas disponibles
from datetime import date

# Convertir números seriales de Excel a fechas datetime
fechas_inicio = pd.to_datetime(df_aperturado['FECHA_INI2'], origin='1899-12-30', unit='D', errors='coerce')
fechas_fin = pd.to_datetime(df_aperturado['FECHA_TERM2'], origin='1899-12-30', unit='D', errors='coerce')

# Filtrar fechas válidas (eliminar NaT y fechas antes de 1900)
fechas_inicio_validas = fechas_inicio[fechas_inicio.notna() & (fechas_inicio.dt.year >= 1900)]
fechas_fin_validas = fechas_fin[fechas_fin.notna() & (fechas_fin.dt.year >= 1900)]

# Establecer valores por defecto seguros
if not fechas_inicio_validas.empty and not fechas_fin_validas.empty:
    fecha_min = fechas_inicio_validas.min().date()
    fecha_max = fechas_fin_validas.max().date()
    anio_min = fechas_inicio_validas.min().year
    anio_max = fechas_fin_validas.max().year
else:
    # Fallback a valores actuales si no hay fechas válidas
    fecha_min = date(2020, 1, 1)
    fecha_max = date(2030, 12, 31)
    anio_min = 2020
    anio_max = 2030

print(f"📅 Rango de fechas detectado: {fecha_min} a {fecha_max}")
print(f"📅 Años disponibles: {anio_min} - {anio_max}")

# Crear widgets con selección múltiple donde se requiere
w_campus_edificio = widgets.SelectMultiple(
    options=campus_edificios,
    value=[campus_edificios[0]] if 'BES' not in campus_edificios else ['BES'],
    description='Edificio Campus:', 
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_campus = widgets.SelectMultiple(
    options=campus_list_options,
    value=[campus_list_options[0]] if 'BES' not in campus_list_options else ['BES'],
    description='Campus Prog:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_periodo = widgets.SelectMultiple(
    options=periodos,
    value=[periodos[0]] if periodos else [],
    description='Periodo Académico:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

# *** NUEVO: Descripción Carrera/Materia con selección múltiple ***
w_desc_carrera_materia = widgets.SelectMultiple(
    options=desc_carreras_materias,
    value=[],
    description='Desc Carrera y/o Materia:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='120px')
)

w_clasificacion_capacidad = widgets.SelectMultiple(
    options=clasificaciones_capacidad,
    value=['EXTERNO'] if 'EXTERNO' in clasificaciones_capacidad else [],
    description='Excluir Clas. (Capacidad):',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_clasificacion_ocupacion = widgets.SelectMultiple(
    options=clasificaciones_ocupacion,
    value=[],
    description='Excluir Clas. (Programación):',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_anio = widgets.Dropdown(
    options=anios,
    value=2025 if 2025 in anios else anios[-1],
    description='Año:',
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

# *** NUEVO: Widgets de rango de fechas ***
w_fecha_inicio = widgets.DatePicker(
    description='Fecha Inicio:',
    value=fecha_min,
    min=date(anio_min, 1, 1),
    max=date(anio_max, 12, 31),
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

w_fecha_fin = widgets.DatePicker(
    description='Fecha Fin:',
    value=fecha_max,
    min=date(anio_min, 1, 1),
    max=date(anio_max, 12, 31),
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

# Checkbox para activar/desactivar filtro de fechas
w_usar_fechas = widgets.Checkbox(
    value=False,
    description='Filtrar por rango de fechas',
    style={'description_width': 'initial'},
    layout=Layout(width='300px')
)

# Botón para actualizar
w_boton = widgets.Button(
    description='🔄 Actualizar Heatmap',
    button_style='primary',
    layout=Layout(width='300px', height='40px')
)

# Botón para descargar CSV
w_boton_csv = widgets.Button(
    description='💾 Descargar CSV Filtrado',
    button_style='success',
    layout=Layout(width='300px', height='40px')
)

# Botón para descargar Resumen Ejecutivo PDF
w_boton_pdf = widgets.Button(
    description='📄 Resumen Ejecutivo PDF',
    button_style='warning',
    layout=Layout(width='300px', height='40px')
)

# Output widget para mostrar el gráfico
output = widgets.Output()

def on_button_click(b):
    with output:
        clear_output(wait=True)
        
        # Determinar si se usan las fechas
        fecha_inicio_filtro = w_fecha_inicio.value if w_usar_fechas.value else None
        fecha_fin_filtro = w_fecha_fin.value if w_usar_fechas.value else None
        
        generar_heatmap_interactivo(
            campus_edificio_list=list(w_campus_edificio.value),
            clasificacion_excluir_capacidad=list(w_clasificacion_capacidad.value),
            clasificacion_excluir_ocupacion=list(w_clasificacion_ocupacion.value),
            anio=w_anio.value,
            campus_list=list(w_campus.value),
            periodo_list=list(w_periodo.value),
            desc_carrera_materia_list=list(w_desc_carrera_materia.value),
            fecha_inicio=fecha_inicio_filtro,
            fecha_fin=fecha_fin_filtro
        )

def on_csv_click(b):
    try:
        if 'df_ocupacion_filtrado' in globals():
            timestamp = pd.Timestamp.now().strftime('%Y%m%d_%H%M%S')
            filename = f'datos_filtrados_{timestamp}.csv'
            df_ocupacion_filtrado.to_csv(filename, index=False, encoding='utf-8-sig')
            print(f"✅ Archivo guardado: {filename}")
            print(f"   Total de registros: {len(df_ocupacion_filtrado):,}")
        else:
            print("⚠️ Primero debes generar el heatmap presionando '🔄 Actualizar Heatmap'")
    except Exception as e:
        print(f"❌ Error al guardar CSV: {str(e)}")

def on_pdf_heatmap_click(b):
    try:
        if 'df_ocupacion_filtrado' in globals() and df_ocupacion_filtrado is not None:
            filtros = {
                'anio': w_anio.value,
                'campus_list': list(w_campus.value),
                'campus_edificio_list': list(w_campus_edificio.value),
                'periodo_list': list(w_periodo.value),
                'desc_carrera_materia_list': list(w_desc_carrera_materia.value),
                'fecha_inicio': w_fecha_inicio.value if w_usar_fechas.value else None,
                'fecha_fin': w_fecha_fin.value if w_usar_fechas.value else None,
                'clasificacion_excluir_capacidad': list(w_clasificacion_capacidad.value),
                'clasificacion_excluir_ocupacion': list(w_clasificacion_ocupacion.value),
            }
            generar_resumen_ejecutivo_pdf(df_ocupacion_filtrado, df3, df2, filtros, tipo_grafico="Heatmap")
        else:
            print("⚠️ Primero debes generar el heatmap presionando '🔄 Actualizar Heatmap'")
    except Exception as e:
        print(f"❌ Error al generar PDF: {str(e)}")
        import traceback
        traceback.print_exc()

w_boton.on_click(on_button_click)
w_boton_csv.on_click(on_csv_click)
w_boton_pdf.on_click(on_pdf_heatmap_click)

# Layout de los controles - Reorganizado en 3 columnas
controles = VBox([
    widgets.HTML("<h3>🎛️ Filtros de Datos</h3>"),
    widgets.HTML("<p style='color: #666; font-size: 12px;'>💡 Usa Ctrl+Click para selección múltiple en campos habilitados</p>"),
    HBox([
        VBox([
            widgets.HTML("<b>Filtros Generales</b>"),
            w_anio,
            w_campus_edificio,
            w_campus
        ]),
        VBox([
            widgets.HTML("<b>Filtros de Programación</b>"),
            w_periodo,
            w_desc_carrera_materia,
            widgets.HTML("<br><b>Rango de Fechas</b>"),
            w_usar_fechas,
            w_fecha_inicio,
            w_fecha_fin
        ]),
        VBox([
            widgets.HTML("<b>Clasificaciones</b>"),
            w_clasificacion_capacidad,
            w_clasificacion_ocupacion
        ])
    ]),
    HBox([w_boton, w_boton_csv, w_boton_pdf]),
    widgets.HTML("<hr>")
])

# Mostrar interfaz
display(controles, output)

# Generar gráfico inicial
with output:
    generar_heatmap_interactivo(
        campus_edificio_list=list(w_campus_edificio.value),
        clasificacion_excluir_capacidad=list(w_clasificacion_capacidad.value),
        clasificacion_excluir_ocupacion=list(w_clasificacion_ocupacion.value),
        anio=w_anio.value,
        campus_list=list(w_campus.value),
        periodo_list=list(w_periodo.value),
        desc_carrera_materia_list=list(w_desc_carrera_materia.value),
        fecha_inicio=None,
        fecha_fin=None
    )
    
```

```python
######### GRÁFICO SUNBURST V5 - OCUPACIÓN POR CAMPUS/EDIFICIO/RANGO/SALA #####################
# Este código debe ejecutarse en una celda separada DESPUÉS del código del heatmap
# Requiere que df3 y df_aperturado ya estén cargados en el notebook
# 
# JERARQUÍA: CAMPUS_AUX (centro) → EDIFICIO → Rango_Capacidad → SALA
# 
# CORRECCIÓN: Usar branchvalues='total' con valores que sumen correctamente


######### FUNCIÓN PRINCIPAL PARA GENERAR SUNBURST #####################

def generar_sunburst_interactivo(
    campus_edificio_list,
    clasificacion_excluir_capacidad,
    clasificacion_excluir_ocupacion,
    anio,
    campus_list,
    periodo_list,
    desc_carrera_materia_list,
    fecha_inicio,
    fecha_fin,
    bloque_inicio,
    bloque_fin,
    dias_sesion_list
):
    """
    Genera el gráfico Sunburst con los filtros seleccionados.
    
    Jerarquía: CAMPUS_AUX (centro) → EDIFICIO → Rango_Capacidad → SALA
    """
    
    print("🔄 Iniciando generación del Sunburst...")
    
    ## 1.- PREPARAR DATOS DE CAPACIDAD (desde df3)
    
    # Filtrar df3 por edificios seleccionados
    filtro_df3 = df3['Campus Edificio'].isin(campus_edificio_list)
    
    # Excluir clasificaciones seleccionadas
    if clasificacion_excluir_capacidad:
        filtro_df3 = filtro_df3 & (~df3['CLASIFICACIÓN'].isin(clasificacion_excluir_capacidad))
    
    # Excluir campus ONL y valores inválidos
    filtro_df3 = filtro_df3 & (df3['Campus Edificio'] != 'ONL') & (df3['Campus Edificio'] != 0)
    
    df3_filtrado = df3[filtro_df3].copy()
    
    print(f"   📁 Registros en df3 después de filtros: {len(df3_filtrado)}")
    
    if df3_filtrado.empty:
        print("⚠️ No hay datos de capacidad con los filtros seleccionados")
        return
    
    # Crear DataFrame de capacidades por sala
    df_capacidades_sala = df3_filtrado.groupby(
        ['Campus Edificio', 'Código Edificio', 'Número Sala']
    ).agg({
        'Capacidad Máxima Sala': 'sum'
    }).reset_index()
    df_capacidades_sala.columns = ['CAMPUS_AUX', 'EDIFICIO', 'SALA', 'CAPACIDAD_SALA']
    
    # Convertir a string para asegurar match
    df_capacidades_sala['CAMPUS_AUX'] = df_capacidades_sala['CAMPUS_AUX'].astype(str)
    df_capacidades_sala['EDIFICIO'] = df_capacidades_sala['EDIFICIO'].astype(str)
    df_capacidades_sala['SALA'] = df_capacidades_sala['SALA'].astype(str)
    
    ## 2.- PREPARAR DATOS DE OCUPACIÓN
    
    df_ocupacion = df_aperturado[
        (df_aperturado['REG_UNICO_LLAV2'] == 1) & 
        (df_aperturado['DIA_SESION'].notna()) &
        (df_aperturado['DIA_SESION'] != '') &
        (df_aperturado['SALA'] != '')
    ].copy()
    
    # Convertir a string
    df_ocupacion['CAMPUS_AUX'] = df_ocupacion['CAMPUS_AUX'].astype(str)
    df_ocupacion['EDIFICIO'] = df_ocupacion['EDIFICIO'].astype(str)
    df_ocupacion['SALA'] = df_ocupacion['SALA'].astype(str)
    
    # Aplicar filtros
    df_ocupacion = df_ocupacion[
        (df_ocupacion['AÑO'] == anio) & 
        (df_ocupacion['CAMPUS'].isin(campus_list))
    ]
    
    df_ocupacion = df_ocupacion[df_ocupacion['CAMPUS_AUX'] != 'ONL']
    
    if campus_edificio_list:
        df_ocupacion = df_ocupacion[df_ocupacion['CAMPUS_AUX'].isin([str(c) for c in campus_edificio_list])]
    
    if periodo_list:
        df_ocupacion = df_ocupacion[df_ocupacion['PERIODO'].isin(periodo_list)]
    
    if desc_carrera_materia_list:
        df_ocupacion = df_ocupacion[df_ocupacion['DESCRIPCIÓN CARRERA_MATERIA'].isin(desc_carrera_materia_list)]
    
    if fecha_inicio and fecha_fin:
        if 'FECHA_INI2' in df_ocupacion.columns and 'FECHA_TERM2' in df_ocupacion.columns:
            df_ocupacion['FECHA_INI2'] = pd.to_datetime(df_ocupacion['FECHA_INI2'], origin='1899-12-30', unit='D', errors='coerce')
            df_ocupacion['FECHA_TERM2'] = pd.to_datetime(df_ocupacion['FECHA_TERM2'], origin='1899-12-30', unit='D', errors='coerce')
            fecha_inicio_dt = pd.to_datetime(fecha_inicio)
            fecha_fin_dt = pd.to_datetime(fecha_fin)
            df_ocupacion = df_ocupacion[
                (df_ocupacion['FECHA_INI2'] <= fecha_fin_dt) & 
                (df_ocupacion['FECHA_TERM2'] >= fecha_inicio_dt)
            ]
    
    if clasificacion_excluir_ocupacion:
        df_ocupacion = df_ocupacion[~df_ocupacion['CLASIFICACIÓN'].isin(clasificacion_excluir_ocupacion)]
    
    if dias_sesion_list:
        df_ocupacion = df_ocupacion[df_ocupacion['DIA_SESION'].isin(dias_sesion_list)]
    
    if bloque_inicio and bloque_fin:
        df_ocupacion['HR_INICIO_AUX_STR'] = df_ocupacion['HR_INICIO_AUX_STR'].astype(str)
        df_ocupacion['HR_FIN_AUX_STR'] = df_ocupacion['HR_FIN_AUX_STR'].astype(str)
        df_ocupacion = df_ocupacion[
            (df_ocupacion['HR_INICIO_AUX_STR'] >= bloque_inicio) & 
            (df_ocupacion['HR_FIN_AUX_STR'] <= bloque_fin)
        ]
    
    print(f"   📁 Registros de ocupación después de filtros: {len(df_ocupacion)}")
    
    if df_ocupacion.empty:
        print("⚠️ No hay datos de ocupación con los filtros seleccionados")
        return
    
    if 'Rango_Capacidad' not in df_ocupacion.columns:
        print("⚠️ El campo 'Rango_Capacidad' no existe en los datos.")
        return
    
    # Guardar DataFrame filtrado como variable global para exportación
    global df_sunburst_filtrado
    df_sunburst_filtrado = df_ocupacion.copy()
    
    ## 3.- CALCULAR MÉTRICAS POR SALA (nivel más bajo)
    
    metricas_sala = df_ocupacion.groupby(
        ['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad', 'SALA']
    ).agg({
        'OCUPACIÓN_SALA': 'mean',
        'INSCRITOS': ['sum', 'max', 'min']
    }).reset_index()
    metricas_sala.columns = ['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad', 'SALA', 
                              'TIP', 'INSCRITOS_TOTAL', 'INSCRITOS_MAX', 'INSCRITOS_MIN']
    
    # Merge con capacidad de sala
    metricas_sala = metricas_sala.merge(
        df_capacidades_sala[['CAMPUS_AUX', 'EDIFICIO', 'SALA', 'CAPACIDAD_SALA']],
        on=['CAMPUS_AUX', 'EDIFICIO', 'SALA'],
        how='left'
    )
    metricas_sala['CAPACIDAD_SALA'] = metricas_sala['CAPACIDAD_SALA'].fillna(30)
    metricas_sala['VALUE'] = metricas_sala['CAPACIDAD_SALA']
    
    # Texto de inscritos MAX-MIN
    metricas_sala['INSCRITOS_RANGO'] = metricas_sala['INSCRITOS_MAX'].astype(int).astype(str) + '-' + \
                                        metricas_sala['INSCRITOS_MIN'].astype(int).astype(str)
    
    print(f"   📊 Salas con métricas: {len(metricas_sala)}")
    
    ## 4.- CALCULAR MÉTRICAS AGREGADAS (promedio directo desde registros originales)
    
    # Nivel RANGO: Promedio directo desde df_ocupacion
    metricas_rango = df_ocupacion.groupby(['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad']).agg({
        'OCUPACIÓN_SALA': 'mean',  # Promedio directo de todos los registros del rango
        'SALA': 'nunique'  # Conteo de salas para hover
    }).reset_index()
    metricas_rango.columns = ['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad', 'TIP', 'N_SALAS']
    
    # Calcular VALUE como suma de capacidades de salas del rango
    value_rango = metricas_sala.groupby(['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad'])['VALUE'].sum().reset_index()
    value_rango.columns = ['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad', 'VALUE']
    metricas_rango = metricas_rango.merge(value_rango, on=['CAMPUS_AUX', 'EDIFICIO', 'Rango_Capacidad'], how='left')
    metricas_rango['VALUE'] = metricas_rango['VALUE'].fillna(1)
    
    print(f"   📊 Rangos con métricas: {len(metricas_rango)}")
    
    # Nivel EDIFICIO: Promedio directo desde df_ocupacion
    metricas_edificio = df_ocupacion.groupby(['CAMPUS_AUX', 'EDIFICIO']).agg({
        'OCUPACIÓN_SALA': 'mean',  # Promedio directo de todos los registros del edificio
        'INSCRITOS': 'sum'
    }).reset_index()
    metricas_edificio.columns = ['CAMPUS_AUX', 'EDIFICIO', 'TIP', 'INSCRITOS_EDIFICIO']
    
    # Calcular VALUE como suma de capacidades de rangos del edificio
    value_edificio = metricas_rango.groupby(['CAMPUS_AUX', 'EDIFICIO'])['VALUE'].sum().reset_index()
    value_edificio.columns = ['CAMPUS_AUX', 'EDIFICIO', 'VALUE']
    metricas_edificio = metricas_edificio.merge(value_edificio, on=['CAMPUS_AUX', 'EDIFICIO'], how='left')
    metricas_edificio['VALUE'] = metricas_edificio['VALUE'].fillna(1)
    
    # Agregar capacidad del edificio desde df3
    capacidad_edificio = df_capacidades_sala.groupby(['CAMPUS_AUX', 'EDIFICIO'])['CAPACIDAD_SALA'].sum().reset_index()
    capacidad_edificio.columns = ['CAMPUS_AUX', 'EDIFICIO', 'CAPACIDAD_EDIFICIO']
    metricas_edificio = metricas_edificio.merge(capacidad_edificio, on=['CAMPUS_AUX', 'EDIFICIO'], how='left')
    metricas_edificio['CAPACIDAD_EDIFICIO'] = metricas_edificio['CAPACIDAD_EDIFICIO'].fillna(metricas_edificio['VALUE'])
    
    print(f"   📊 Edificios con métricas: {len(metricas_edificio)}")
    
    # Nivel CAMPUS: Promedio directo desde df_ocupacion
    metricas_campus = df_ocupacion.groupby('CAMPUS_AUX').agg({
        'OCUPACIÓN_SALA': 'mean',  # Promedio directo de todos los registros del campus
        'INSCRITOS': 'sum'
    }).reset_index()
    metricas_campus.columns = ['CAMPUS_AUX', 'TIP', 'INSCRITOS_CAMPUS']
    
    # Calcular VALUE como suma de capacidades de edificios del campus
    value_campus = metricas_edificio.groupby('CAMPUS_AUX')['VALUE'].sum().reset_index()
    value_campus.columns = ['CAMPUS_AUX', 'VALUE']
    metricas_campus = metricas_campus.merge(value_campus, on='CAMPUS_AUX', how='left')
    metricas_campus['VALUE'] = metricas_campus['VALUE'].fillna(1)
    
    # Agregar capacidad del campus desde df3
    capacidad_campus = df_capacidades_sala.groupby('CAMPUS_AUX')['CAPACIDAD_SALA'].sum().reset_index()
    capacidad_campus.columns = ['CAMPUS_AUX', 'CAPACIDAD_CAMPUS']
    metricas_campus = metricas_campus.merge(capacidad_campus, on='CAMPUS_AUX', how='left')
    metricas_campus['CAPACIDAD_CAMPUS'] = metricas_campus['CAPACIDAD_CAMPUS'].fillna(metricas_campus['VALUE'])
    
    print(f"   📊 Campus con métricas: {len(metricas_campus)}")
    
    ## 5.- CONSTRUIR LISTAS PARA SUNBURST
    
    print("   🎨 Construyendo gráfico...")
    
    ids = []
    labels = []
    parents = []
    values = []
    colors = []
    hover_texts = []
    
    # Nivel CAMPUS (centro)
    for _, row in metricas_campus.iterrows():
        campus = str(row['CAMPUS_AUX'])
        ids.append(campus)
        labels.append(campus)
        parents.append('')
        values.append(float(row['VALUE']))
        colors.append(float(row['TIP']) if pd.notna(row['TIP']) else 0)
        ocupacion_pct = (row['TIP'] * 100) if pd.notna(row['TIP']) else 0
        hover_texts.append(
            f"<b>{campus}</b><br>" +
            f"Ocupación promedio: {ocupacion_pct:.2f}%"
        )
    
    # Nivel EDIFICIO
    for _, row in metricas_edificio.iterrows():
        campus = str(row['CAMPUS_AUX'])
        edificio = str(row['EDIFICIO'])
        id_edificio = f"{campus} - {edificio}"
        ids.append(id_edificio)
        labels.append(edificio)
        parents.append(campus)
        values.append(float(row['VALUE']))
        colors.append(float(row['TIP']) if pd.notna(row['TIP']) else 0)
        ocupacion_pct = (row['TIP'] * 100) if pd.notna(row['TIP']) else 0
        hover_texts.append(
            f"<b>{edificio}</b><br>" +
            f"Ocupación promedio: {ocupacion_pct:.2f}%"
        )
    
    # Nivel RANGO_CAPACIDAD
    for _, row in metricas_rango.iterrows():
        campus = str(row['CAMPUS_AUX'])
        edificio = str(row['EDIFICIO'])
        rango = str(row['Rango_Capacidad'])
        id_parent = f"{campus} - {edificio}"
        id_rango = f"{campus} - {edificio} - {rango}"
        ids.append(id_rango)
        labels.append(rango)
        parents.append(id_parent)
        values.append(float(row['VALUE']))
        colors.append(float(row['TIP']) if pd.notna(row['TIP']) else 0)
        ocupacion_pct = (row['TIP'] * 100) if pd.notna(row['TIP']) else 0
        n_salas = row['N_SALAS']
        hover_texts.append(
            f"<b>{rango}</b><br>" +
            f"Ocupación promedio: {ocupacion_pct:.2f}%<br>" +
            f"Salas: {n_salas:.0f}"
        )
    
    # Nivel SALA
    for _, row in metricas_sala.iterrows():
        campus = str(row['CAMPUS_AUX'])
        edificio = str(row['EDIFICIO'])
        rango = str(row['Rango_Capacidad'])
        sala = str(row['SALA'])
        id_parent = f"{campus} - {edificio} - {rango}"
        id_sala = f"{campus} - {edificio} - {rango} - {sala}"
        ids.append(id_sala)
        labels.append(f"Sala {sala}")
        parents.append(id_parent)
        values.append(float(row['VALUE']))
        colors.append(float(row['TIP']) if pd.notna(row['TIP']) else 0)
        ocupacion_pct = (row['TIP'] * 100) if pd.notna(row['TIP']) else 0
        inscritos_rango = row['INSCRITOS_RANGO']
        capacidad = row['CAPACIDAD_SALA'] if pd.notna(row['CAPACIDAD_SALA']) else 0
        hover_texts.append(
            f"<b>Sala {sala}</b><br>" +
            f"Inscritos: {inscritos_rango}<br>" +
            f"Capacidad: {capacidad:,.0f}<br>" +
            f"Ocupación: {ocupacion_pct:.2f}%"
        )
    
    print(f"   📊 Total elementos en Sunburst: {len(ids)}")
    
    ## 6.- CREAR GRÁFICO
    
    fig = go.Figure(go.Sunburst(
        ids=ids,
        labels=labels,
        parents=parents,
        values=values,
        marker=dict(
            colors=colors,
            colorscale='RdYlGn',##### CAMBIO DE PALETA #################################################################           ##
            cmin=0,
            cmax=1,
            colorbar=dict(
                title="Ocupación",
                tickformat=".0%",
                tickvals=[0, 0.25, 0.5, 0.75, 1],
                ticktext=['0%', '25%', '50%', '75%', '100%']
            )
        ),
        hovertext=hover_texts,
        hoverinfo='text',
        branchvalues='total'  # Volvemos a total, ahora los valores suman correctamente
    ))
    
    # Título dinámico - Mostrar códigos de campus, no conteo
    campus_str = ', '.join([str(c) for c in campus_list])
    
    titulo = f'<b>Ocupación por Campus - Vista Sunburst</b><br>' + \
             f'<sub>Campus: {campus_str} | Año: {anio}</sub>'
    
    if bloque_inicio and bloque_fin:
        titulo += f'<br><sub>Bloques: {bloque_inicio} - {bloque_fin}</sub>'
    
    if dias_sesion_list:
        dias_str = ', '.join(dias_sesion_list) if len(dias_sesion_list) <= 4 else f"{len(dias_sesion_list)} días"
        titulo += f'<br><sub>Días: {dias_str}</sub>'
    
    fig.update_layout(
        title=dict(
            text=titulo,
            x=0.5,
            xanchor='center',
            y=0.98,
            yanchor='top'
        ),
        width=950,
        height=900,
        font=dict(size=12),
        margin=dict(t=150, l=50, r=50, b=50)
    )
    
    fig.show()
    
    ## 7.- MOSTRAR RESUMEN
    print(f"\n📊 Resumen de datos Sunburst:")
    print(f"   • Campus incluidos: {metricas_campus['CAMPUS_AUX'].nunique()}")
    print(f"   • Edificios incluidos: {metricas_edificio['EDIFICIO'].nunique()}")
    print(f"   • Rangos de capacidad: {metricas_rango['Rango_Capacidad'].nunique()}")
    print(f"   • Salas incluidas: {len(metricas_sala)}")
    print(f"   • Total registros procesados: {len(df_ocupacion):,}")
    
    print(f"\n📍 Ocupación promedio por Campus:")
    for _, row in metricas_campus.iterrows():
        campus = row['CAMPUS_AUX']
        tip = (row['TIP'] * 100) if pd.notna(row['TIP']) else 0
        capacidad = row['CAPACIDAD_CAMPUS'] if pd.notna(row['CAPACIDAD_CAMPUS']) else 0
        inscritos = row['INSCRITOS_CAMPUS'] if pd.notna(row['INSCRITOS_CAMPUS']) else 0
        print(f"   • {campus}: {tip:.2f}% | Capacidad: {capacidad:,.0f} | Inscritos: {inscritos:,.0f}")


######### CREAR WIDGETS DE FILTRO PARA SUNBURST #####################

print("🌀 Configurando filtros para gráfico Sunburst...")

# Obtener valores únicos para los filtros
sunburst_campus_edificios = sorted([str(c) for c in df3['Campus Edificio'].dropna().unique() if c != 0 and str(c) != 'ONL'])

sunburst_clasificaciones_capacidad = sorted(df3['CLASIFICACIÓN'].dropna().unique(), key=str)
sunburst_clasificaciones_ocupacion = sorted(df_aperturado['CLASIFICACIÓN'].dropna().unique(), key=str)
sunburst_anios = sorted(df_aperturado['AÑO'].dropna().unique(), key=str)
sunburst_campus_list_options = sorted(df_aperturado['CAMPUS'].dropna().unique(), key=str)
sunburst_periodos = sorted(df_aperturado['PERIODO'].dropna().unique(), key=str)
sunburst_desc_carreras_materias = sorted(df_aperturado['DESCRIPCIÓN CARRERA_MATERIA'].dropna().unique(), key=str)

# Obtener valores únicos para bloques (horas)
sunburst_bloques_inicio = sorted(df_aperturado['HR_INICIO_AUX_STR'].dropna().astype(str).unique(), key=str)
sunburst_bloques_fin = sorted(df_aperturado['HR_FIN_AUX_STR'].dropna().astype(str).unique(), key=str)

# Obtener días de sesión
sunburst_dias = ['LUNES', 'MARTES', 'MIERCOLES', 'JUEVES', 'VIERNES', 'SABADO', 'DOMINGO']
dias_disponibles = [d for d in sunburst_dias if d in df_aperturado['DIA_SESION'].unique()]

# Obtener rango de fechas
from datetime import date

fechas_inicio_sb = pd.to_datetime(df_aperturado['FECHA_INI2'], origin='1899-12-30', unit='D', errors='coerce')
fechas_fin_sb = pd.to_datetime(df_aperturado['FECHA_TERM2'], origin='1899-12-30', unit='D', errors='coerce')

fechas_inicio_validas_sb = fechas_inicio_sb[fechas_inicio_sb.notna() & (fechas_inicio_sb.dt.year >= 1900)]
fechas_fin_validas_sb = fechas_fin_sb[fechas_fin_sb.notna() & (fechas_fin_sb.dt.year >= 1900)]

if not fechas_inicio_validas_sb.empty and not fechas_fin_validas_sb.empty:
    fecha_min_sb = fechas_inicio_validas_sb.min().date()
    fecha_max_sb = fechas_fin_validas_sb.max().date()
    anio_min_sb = fechas_inicio_validas_sb.min().year
    anio_max_sb = fechas_fin_validas_sb.max().year
else:
    fecha_min_sb = date(2020, 1, 1)
    fecha_max_sb = date(2030, 12, 31)
    anio_min_sb = 2020
    anio_max_sb = 2030

# Crear widgets
w_sb_campus_edificio = widgets.SelectMultiple(
    options=sunburst_campus_edificios,
    value=[sunburst_campus_edificios[0]] if sunburst_campus_edificios else [],
    description='Edificio Campus:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_sb_campus = widgets.SelectMultiple(
    options=sunburst_campus_list_options,
    value=[sunburst_campus_list_options[0]] if sunburst_campus_list_options else [],
    description='Campus Prog:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_sb_periodo = widgets.SelectMultiple(
    options=sunburst_periodos,
    value=[sunburst_periodos[0]] if sunburst_periodos else [],
    description='Periodo Académico:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_sb_desc_carrera_materia = widgets.SelectMultiple(
    options=sunburst_desc_carreras_materias,
    value=[],
    description='Desc Carrera/Materia:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='120px')
)

w_sb_clasificacion_capacidad = widgets.SelectMultiple(
    options=sunburst_clasificaciones_capacidad,
    value=['EXTERNO'] if 'EXTERNO' in sunburst_clasificaciones_capacidad else [],
    description='Excluir Clas. (Cap):',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_sb_clasificacion_ocupacion = widgets.SelectMultiple(
    options=sunburst_clasificaciones_ocupacion,
    value=[],
    description='Excluir Clas. (Prog):',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='100px')
)

w_sb_anio = widgets.Dropdown(
    options=sunburst_anios,
    value=2025 if 2025 in sunburst_anios else sunburst_anios[-1] if sunburst_anios else None,
    description='Año:',
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

w_sb_bloque_inicio = widgets.Dropdown(
    options=[''] + sunburst_bloques_inicio,
    value='',
    description='Bloque Inicio:',
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

w_sb_bloque_fin = widgets.Dropdown(
    options=[''] + sunburst_bloques_fin,
    value='',
    description='Bloque Fin:',
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

w_sb_dias = widgets.SelectMultiple(
    options=dias_disponibles,
    value=dias_disponibles,
    description='Días Sesión:',
    style={'description_width': '150px'},
    layout=Layout(width='300px', height='120px')
)

w_sb_fecha_inicio = widgets.DatePicker(
    description='Fecha Inicio:',
    value=fecha_min_sb,
    min=date(anio_min_sb, 1, 1),
    max=date(anio_max_sb, 12, 31),
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

w_sb_fecha_fin = widgets.DatePicker(
    description='Fecha Fin:',
    value=fecha_max_sb,
    min=date(anio_min_sb, 1, 1),
    max=date(anio_max_sb, 12, 31),
    style={'description_width': '150px'},
    layout=Layout(width='300px')
)

w_sb_usar_fechas = widgets.Checkbox(
    value=False,
    description='Filtrar por rango de fechas',
    style={'description_width': 'initial'},
    layout=Layout(width='300px')
)

w_sb_boton = widgets.Button(
    description='🌀 Actualizar Sunburst',
    button_style='primary',
    layout=Layout(width='300px', height='40px')
)

# Botón para descargar CSV
w_sb_boton_csv = widgets.Button(
    description='💾 Descargar CSV Filtrado',
    button_style='success',
    layout=Layout(width='300px', height='40px')
)

# Botón para descargar Resumen Ejecutivo PDF
w_sb_boton_pdf = widgets.Button(
    description='📄 Resumen Ejecutivo PDF',
    button_style='warning',
    layout=Layout(width='300px', height='40px')
)

output_sunburst = widgets.Output()

# Variable global para almacenar el DataFrame filtrado
df_sunburst_filtrado = None

def on_sunburst_button_click(b):
    with output_sunburst:
        clear_output(wait=True)
        
        fecha_inicio_filtro = w_sb_fecha_inicio.value if w_sb_usar_fechas.value else None
        fecha_fin_filtro = w_sb_fecha_fin.value if w_sb_usar_fechas.value else None
        
        bloque_inicio = w_sb_bloque_inicio.value if w_sb_bloque_inicio.value else None
        bloque_fin = w_sb_bloque_fin.value if w_sb_bloque_fin.value else None
        
        generar_sunburst_interactivo(
            campus_edificio_list=list(w_sb_campus_edificio.value),
            clasificacion_excluir_capacidad=list(w_sb_clasificacion_capacidad.value),
            clasificacion_excluir_ocupacion=list(w_sb_clasificacion_ocupacion.value),
            anio=w_sb_anio.value,
            campus_list=list(w_sb_campus.value),
            periodo_list=list(w_sb_periodo.value),
            desc_carrera_materia_list=list(w_sb_desc_carrera_materia.value),
            fecha_inicio=fecha_inicio_filtro,
            fecha_fin=fecha_fin_filtro,
            bloque_inicio=bloque_inicio,
            bloque_fin=bloque_fin,
            dias_sesion_list=list(w_sb_dias.value)
        )

w_sb_boton.on_click(on_sunburst_button_click)

def on_csv_sunburst_click(b):
    """Función para descargar el CSV filtrado del Sunburst"""
    try:
        global df_sunburst_filtrado
        if df_sunburst_filtrado is not None and len(df_sunburst_filtrado) > 0:
            timestamp = pd.Timestamp.now().strftime('%Y%m%d_%H%M%S')
            # Ruta de descarga específica
            ruta_descarga = r'C:\Users\diego.perezp\Downloads'
            filename = f'{ruta_descarga}\\sunburst_datos_filtrados_{timestamp}.csv'
            df_sunburst_filtrado.to_csv(filename, index=False, encoding='utf-8-sig')
            print(f"✅ Archivo guardado: {filename}")
            print(f"   Total de registros: {len(df_sunburst_filtrado):,}")
        else:
            print("⚠️ Primero debes generar el Sunburst presionando '🌀 Actualizar Sunburst'")
    except Exception as e:
        print(f"❌ Error al guardar CSV: {str(e)}")

def on_pdf_sunburst_click(b):
    try:
        global df_sunburst_filtrado
        if df_sunburst_filtrado is not None and len(df_sunburst_filtrado) > 0:
            filtros = {
                'anio': w_sb_anio.value,
                'campus_list': list(w_sb_campus.value),
                'campus_edificio_list': list(w_sb_campus_edificio.value),
                'periodo_list': list(w_sb_periodo.value),
                'desc_carrera_materia_list': list(w_sb_desc_carrera_materia.value),
                'fecha_inicio': w_sb_fecha_inicio.value if w_sb_usar_fechas.value else None,
                'fecha_fin': w_sb_fecha_fin.value if w_sb_usar_fechas.value else None,
                'clasificacion_excluir_capacidad': list(w_sb_clasificacion_capacidad.value),
                'clasificacion_excluir_ocupacion': list(w_sb_clasificacion_ocupacion.value),
                'bloque_inicio': w_sb_bloque_inicio.value if w_sb_bloque_inicio.value else None,
                'bloque_fin': w_sb_bloque_fin.value if w_sb_bloque_fin.value else None,
                'dias_sesion_list': list(w_sb_dias.value),
            }
            generar_resumen_ejecutivo_pdf(df_sunburst_filtrado, df3, df2, filtros, tipo_grafico="Sunburst")
        else:
            print("⚠️ Primero debes generar el Sunburst presionando '🌀 Actualizar Sunburst'")
    except Exception as e:
        print(f"❌ Error al generar PDF: {str(e)}")
        import traceback
        traceback.print_exc()

w_sb_boton_csv.on_click(on_csv_sunburst_click)
w_sb_boton_pdf.on_click(on_pdf_sunburst_click)

controles_sunburst = VBox([
    widgets.HTML("<h3>🌀 Gráfico Sunburst - Ocupación por Campus/Edificio/Rango/Sala</h3>"),
    widgets.HTML("""
        <p style='color: #666; font-size: 12px;'>
        💡 <b>Jerarquía:</b> Centro = Campus → 2º anillo = Edificio → 3º anillo = Rango Capacidad → 4º anillo = Sala<br>
        🎨 <b>Color:</b> Promedio de ocupación (Amarillo = baja, Rojo = alta)<br>
        📏 <b>Tamaño:</b> Basado en capacidad (suma de capacidades de salas)<br>
        👆 Usa Ctrl+Click para selección múltiple
        </p>
    """),
    HBox([
        VBox([
            widgets.HTML("<b>Filtros Generales</b>"),
            w_sb_anio,
            w_sb_campus_edificio,
            w_sb_campus
        ]),
        VBox([
            widgets.HTML("<b>Filtros de Programación</b>"),
            w_sb_periodo,
            w_sb_desc_carrera_materia,
            widgets.HTML("<br><b>Rango de Fechas</b>"),
            w_sb_usar_fechas,
            w_sb_fecha_inicio,
            w_sb_fecha_fin
        ]),
        VBox([
            widgets.HTML("<b>Filtros de Bloques y Días</b>"),
            w_sb_bloque_inicio,
            w_sb_bloque_fin,
            w_sb_dias,
            widgets.HTML("<br><b>Clasificaciones</b>"),
            w_sb_clasificacion_capacidad,
            w_sb_clasificacion_ocupacion
        ])
    ]),
    HBox([w_sb_boton, w_sb_boton_csv, w_sb_boton_pdf]),
    widgets.HTML("<hr>")
])

display(controles_sunburst, output_sunburst)

print("⏳ Generando gráfico inicial...")
with output_sunburst:
    generar_sunburst_interactivo(
        campus_edificio_list=list(w_sb_campus_edificio.value),
        clasificacion_excluir_capacidad=list(w_sb_clasificacion_capacidad.value),
        clasificacion_excluir_ocupacion=list(w_sb_clasificacion_ocupacion.value),
        anio=w_sb_anio.value,
        campus_list=list(w_sb_campus.value),
        periodo_list=list(w_sb_periodo.value),
        desc_carrera_materia_list=list(w_sb_desc_carrera_materia.value),
        fecha_inicio=None,
        fecha_fin=None,
        bloque_inicio=None,
        bloque_fin=None,
        dias_sesion_list=list(w_sb_dias.value)
    )
```
